## Introdução à Visualização de Dados

Este bloco de código foi projetado para carregar dados, realizar uma Análise Exploratória de Dados (EDA) e visualizar as distribuições de variáveis importantes, bem como suas correlações.

### Funções Principais:

1.  **Carregamento de Dados:**
    *   `import pandas as pd`: Importa a biblioteca pandas para manipulação de dados.
    *   `df = pd.read_csv('clientes-v3-preparado.csv')`: Carrega o arquivo CSV `'clientes-v3-preparado.csv'` em um DataFrame pandas chamado `df`.
    *   `print(df.head().to_string())`: Exibe as primeiras linhas do DataFrame para uma visualização rápida dos dados.

2.  **Visualização de Distribuição (Histogramas):**
    *   `import matplotlib.pyplot as plt`: Importa a biblioteca matplotlib para criação de gráficos.
    *   `plt.hist(df['salario'])`: Cria um histograma básico da coluna 'salario'.
    *   `plt.figure(figsize = (10, 6))`, `plt.hist(df['salario'], bins = 100, color = 'green', alpha = 0.8)`: Cria um histograma mais detalhado para 'salario', com tamanho personalizado, 100 barras, cor verde e transparência.
    *   `plt.title`, `plt.xlabel`, `plt.xticks`, `plt.ylabel`, `plt.grid(True)`: Adiciona título, rótulos aos eixos, define os 'ticks' do eixo x para melhor legibilidade do salário, e adiciona uma grade ao gráfico.

3.  **Visualização de Relações (Gráficos de Dispersão):**
    *   `plt.figure(figsize = (10, 6))`: Inicia uma nova figura para os gráficos.
    *   `plt.subplot(2,1,1)`: Cria uma sub-grade de gráficos (2 linhas, 1 coluna) e seleciona a primeira posição.
    *   `plt.scatter(df['salario'], df['salario'])`: Cria um gráfico de dispersão do 'salario' contra ele mesmo (geralmente usado para verificar a distribuição ou identidades).
    *   `plt.subplot(1,2,2)`: Cria uma sub-grade de gráficos (1 linha, 2 colunas) e seleciona a segunda posição.
    *   `plt.scatter(df['salario'], df['anos_experiencia'], color = '#5883a8', alpha = 0.6, s=30)`: Cria um gráfico de dispersão de 'salario' contra 'anos_experiencia', com cor hexadecimal, transparência e tamanho dos pontos.

4.  **Visualização de Correlação (Mapa de Calor):**
    *   `import seaborn as sns`: Importa a biblioteca seaborn para gráficos estatísticos.
    *   `corr = df[['salario', 'anos_experiencia']].corr()`: Calcula a matriz de correlação entre 'salario' e 'anos_experiencia'.
    *   `plt.subplot(2, 2, 3)`: Cria uma sub-grade de gráficos (2 linhas, 2 colunas) e seleciona a terceira posição.
    *   `sns.heatmap(corr, annot=True, cmap = 'coolwarm')`: Gera um mapa de calor da matriz de correlação, exibindo os valores (`annot=True`) e usando um esquema de cores 'coolwarm'.

5.  **Ajuste de Layout:**
    *   `plt.tight_layout()`: Ajusta automaticamente os parâmetros da subtrama para que se encaixem bem na área da figura, evitando sobreposições.
    *   `plt.show()`: Exibe todos os gráficos criados.

## Resumo sobre Dispersão em Estatística

### O que é Dispersão?

Em estatística, **dispersão** (também conhecida como variabilidade ou espalhamento) refere-se ao grau em que os valores de um conjunto de dados se estendem ou se agrupam em torno de uma medida de tendência central (como a média, mediana ou moda).

Quando os dados estão muito próximos uns dos outros, dizemos que a dispersão é baixa. Por outro lado, se os dados estão amplamente espalhados, a dispersão é alta.

### O que se pretende analisar com a Dispersão?

A análise da dispersão visa entender e quantificar a **homogeneidade ou heterogeneidade** de um conjunto de dados. Ela complementa as medidas de tendência central, pois duas distribuições podem ter a mesma média, mas comportamentos muito diferentes em termos de como seus dados estão distribuídos.

Pretende-se analisar:
*   **Consistência/Variabilidade:** Quão consistentes são os dados? Há muita flutuação ou são relativamente estáveis?
*   **Risco/Incerteza:** Em contextos financeiros ou de processos, alta dispersão pode indicar maior risco ou imprevisibilidade.
*   **Qualidade:** Em controle de qualidade, baixa dispersão é geralmente desejável, indicando que um processo está produzindo resultados uniformes.
*   **Comparação entre grupos:** Permite comparar a variabilidade entre diferentes grupos ou conjuntos de dados.

### Qual a função da Dispersão?

A principal função das medidas de dispersão é **fornecer uma imagem mais completa da distribuição dos dados**, indo além do valor central. Elas ajudam a:

1.  **Contextualizar as Medidas de Tendência Central:** Uma média de 100 pode significar coisas muito diferentes se os dados variam de 90 a 110 (baixa dispersão) ou de 0 a 200 (alta dispersão).
2.  **Quantificar a Variabilidade:** Através de indicadores como:
    *   **Amplitude (Range):** Diferença entre o valor máximo e mínimo.
    *   **Desvio Médio:** Média dos desvios absolutos de cada valor em relação à média.
    *   **Variância:** Média dos quadrados dos desvios de cada valor em relação à média (penaliza mais desvios maiores).
    *   **Desvio Padrão:** Raiz quadrada da variância, retornando a medida para a mesma unidade dos dados originais, sendo a mais utilizada.
    *   **Coeficiente de Variação:** Medida relativa de dispersão, útil para comparar a variabilidade de conjuntos de dados com médias muito diferentes.
3.  **Apoiar a Tomada de Decisão:** Ajuda a decidir se um processo está sob controle, se um investimento é muito arriscado, ou se dois grupos são significativamente diferentes.
4.  **Fundamentar Inferências Estatísticas:** É essencial para testes de hipóteses e construção de intervalos de confiança, pois a dispersão afeta a precisão das estimativas.

```python
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('clientes-v3-preparado.csv')
print(df.head().to_string())

# Histograma
plt.hist(df['salario'])
plt.show()

# Histograma - Parâmetros
plt.figure(figsize = (10, 6))
plt.hist(df['salario'], bins = 100, color = 'green', alpha = 0.8)
plt.title('Histograma - Distribuição de salários')
plt.xlabel('Salário')
plt.xticks(ticks=range(0, int(df['salario'].max())+2000, 2000))
plt.ylabel('Frequência')
plt.grid(True)
plt.show()

# Múltiplos gráficos
plt.figure(figsize = (10, 6))
plt.subplot(2,1,1) # 2 Linha, 2 colunas, 1º Gráfico
# Gráfico de Dispersão
plt.scatter(df['salario'], df['salario'])
plt.title('Dispersão - Salario e Salário')
plt.xlabel('Salário')
plt.ylabel('Salário')

plt.subplot(1,2,2) # 1 Linha, 2 Colunas, 2º Gráfico
plt.scatter(df['salario'], df['anos_experiencia'], color = '#5883a8', alpha = 0.6, s=30) # cor hexadecimal online
plt.title('Dispersão - Idade e anos experiência')
plt.xlabel('Salário')
plt.ylabel('Anos experiência')

# Mapa de Calor

corr = df[['salario', 'anos_experiencia']].corr()
plt.subplot(2, 2, 3) # 2 linhas, 2 Colunas, 3º Gráfico
sns.heatmap(corr, annot=True, cmap = 'coolwarm')
plt.title('Correlação Salário e Idade')

plt.tight_layout() # Ajustar espaçamentos
plt.show()
```
