#Tratar dados inconsistentes

Este notebook realiza o tratamento de um conjunto de dados de clientes. As principais etapas incluem:

1.  **Leitura dos Dados:** Carrega o arquivo `clientes_remove_outliers.csv` em um DataFrame Pandas.
2.  **Mascaramento de CPF:** Cria uma coluna mascarada para o CPF (`cpf_mascara`) para proteger dados pessoais.
3.  **Tratamento de Datas:** Converte a coluna 'data' para o formato datetime, corrige datas inválidas e calcula a idade ajustada dos clientes.
4.  **Normalização de Endereços:** Extrai informações como `endereco_curto`, `bairro` e `estado_sigla` da coluna 'endereco', além de verificar a formatação do endereço.
5.  **Validação de Dados:** Corrige CPFs com formato incorreto e valida as siglas de estado contra uma lista de estados brasileiros válidos.
6.  **Criação de DataFrame Final:** Seleciona e renomeia as colunas relevantes (`nome`, `cpf`, `idade`, `data`, `endereco`, `bairro`, `estado`) para criar um novo DataFrame `df_salvar`.
7.  **Salvamento dos Dados:** Salva o DataFrame tratado em um novo arquivo CSV chamado `clientes_tratados.csv`.

```python
import pandas as pd
import numpy as np

from limpeza_dados import df_salvar

pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

df = pd.read_csv('clientes_remove_outliers.csv')
print(df.head())

# Mascarar dados pessoais
df['cpf_mascara'] = df['cpf'].apply(lambda cpf: f'{cpf[:3]}.***.***-{cpf[-2]}')
print(df.head())

# Corrigir Datas
df['data'] = pd.to_datetime(df['data'], format='%Y-%m-%d', errors='coerce')

data_atual = pd.to_datetime('today')
df['data_atualizada'] = df['data'].where(df['data'] <= data_atual, pd.to_datetime('1900-01-01'))
df['idade_ajustada'] = data_atual.year - df['data_atualizada'].dt.year
df['idade_ajustada'] -= ((data_atual.month <= df['data_atualizada'].dt.month) & (data_atual.day < df['data_atualizada'].dt.day)).astype(int)
df.loc[df['idade_ajustada'] > 100, 'idade_ajustada'] = np.nan

# Corrigir campos com múltiplas informações
df['endereco_curto'] = df['endereco'].apply(lambda x: x.split('\n')[0].strip())
df['bairro'] = df['endereco'].apply(lambda x: x.split('\n')[1].strip() if len(x.split('\n')) > 1 else 'Desconhecido')
df['estado_sigla'] = df['endereco'].apply(lambda x: x.split(' / ')[-1].strip() if len(x.split('\n')) > 1 else 'Desconhecido')

# Verificar formatação do endereço
df['endereco_curto'] = df['endereco_curto'].apply(lambda x: 'Endereço inválido' if len(x) > 50 or len(x) < 50 else x)

# Corrigir dados errôneos
df['cpf'] = df['cpf'].apply(lambda x: x if len(x) == 14 else 'CPF inválido')

estados_br = ['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO']
df['estado_sigla'] = df['estado_sigla'].str.upper().apply(lambda x: x if x in estados_br else 'Desconhecido')

print('Dados tratados:\n', df.head())

df['cpf'] = df['cpf_mascara']
df['idade'] = df['idade_ajustada']
df['endereco'] = df['endereco_curto']
df['estado'] = df['estado_sigla']
df_salvar = df[['nome', 'cpf', 'idade', 'data', 'endereco', 'bairro', 'estado']]
df_salvar.to_csv('clientes_tratados.csv', index=False)

print('Novo Dataframe:\n', pd.read_csv('clientes_tratados.csv'))
```

## Resumo das Bibliotecas Principais e Comandos Comuns

### Pandas
A biblioteca `pandas` é uma ferramenta fundamental em Python para manipulação e análise de dados. Ela oferece estruturas de dados eficientes e flexíveis, como o DataFrame, que facilitam o trabalho com dados tabulares. Seus principais comandos incluem:
*   `pd.read_csv()`: Utilizado para carregar dados de arquivos CSV em um DataFrame.
*   `df.head()`: Exibe as primeiras N linhas do DataFrame, útil para uma inspeção rápida dos dados.
*   `df['coluna']`: Permite selecionar uma coluna específica do DataFrame.
*   `df.apply()`: Aplica uma função ao longo de um eixo (linhas ou colunas) de um DataFrame.
*   `pd.to_datetime()`: Converte uma coluna para o formato de data e hora, essencial para tratamento de datas.
*   `df.to_csv()`: Salva um DataFrame em um arquivo CSV.

### NumPy
A biblioteca `numpy` é a base para computação numérica em Python. Ela fornece suporte para arrays e matrizes multidimensionais de alto desempenho, além de uma vasta coleção de funções matemáticas para operar sobre esses arrays. Comandos comuns incluem:
*   `np.array()`: Cria um array NumPy a partir de uma lista ou outra sequência.
*   `np.nan`: Representa um valor "Not a Number" (Não é um Número), frequentemente usado para indicar dados ausentes ou inválidos.
*   Operações numéricas básicas: Funções como `np.sqrt()`, `np.sum()`, `np.mean()`, entre outras, para realizar cálculos eficientes em arrays.
