```python
# if/else
num1 = 10
num2 = 50
if num1 > num2:
  print("num1 é o maior número")
else:
  print("num2 é o maior número")
```

```python
# if / elif / else
num1 = 10
num2 = 50
if num1 > num2:
  print(" O num1 é o maior número")
elif num2 < num1:
  print( "O num2 é o menor número")
elif num2 > num1:
  print("O num2 é maior que o num1")
else:
  print("Os números são iguais")
```
