### CODIFICAÇÃO E VARIÁVEIS CATEGÓRICAS

Este código Python realiza diversas operações de pré-processamento em um DataFrame pandas para preparar os dados para análise ou modelagem.

1.  **Carregamento de Dados**: Ele carrega um arquivo CSV chamado `'dados/clientes-v2-tratados.csv'` em um DataFrame `df`.
2.  **Codificação One-Hot para 'estado_civil'**: Converte a coluna categórica `'estado_civil'` em várias colunas binárias (one-hot encoding).
3.  **Codificação Ordinal para 'nivel_educação'**: Mapeia os valores da coluna categórica `'nivel_educacao'` para uma ordem numérica específica (Ensino Fundamental = 1, Ensino Médio = 2, etc.).
4.  **Codificação Categórica para 'area_atuacao'**: Transforma a coluna `'area_atuacao'` em códigos numéricos.
5.  **Label Encoding para 'estado'**: Utiliza o `LabelEncoder` para converter os valores da coluna categórica `'estado'` em números inteiros.

Em essência, o código está preparando dados categóricos para que possam ser usados em algoritmos de machine learning.

```python
# Análise Exploratória de Dados (AED)
import pandas as pd


df = pd.read_csv('dados/clientes-v2.csv')

print(df.head().to_string())
print(df.tail().to_string())
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y', errors='coerce') # Pode ser tratado na preparação

print('Verificação inicial: ')
print(df.info())

print('Analise de dados nulos:\n', df.isnull().sum())
print('% de dados nulos:\n', df.isnull().mean() * 100)

df.dropna(inplace=True) # Deletar todos campos nulos
print('Confirma remoção de dados nulos:\n', df.isnull().sum().sum()) # confirmar se foi deletado todos dados nulos

print('Analise de dados duplicados:\n', df.duplicated().sum())

print('Analise de dados únicos:\n', df.nunique())

print('Estatística dos dados:\n', df.describe()) # Estatística dos dados, quartil, std(desvio padrão)

# Exclui dados sensíveis - identificar usuário (CPF, endereço) e pais(ja que é so Brasil), cria novo df
df= df[['idade', 'data', 'estado', 'salario', 'nivel_educacao', 'numero_filhos', 'estado_civil', 'area_atuacao']]
print(df.head().to_string())

df.to_csv('dados/clientes-v2-tratados.csv', index=False)
```

```python
import pandas as pd # Importa a biblioteca pandas para manipulação e análise de dados tabulares.
from sklearn.preprocessing import LabelEncoder # Importa LabelEncoder do scikit-learn para codificação de rótulos (transformar categorias em números inteiros).

pd.set_option('display.width', None) # Configura o pandas para exibir todas as colunas de um DataFrame, sem truncamento.

df = pd.read_csv('dados/clientes-v2-tratados.csv') # Carrega o arquivo CSV 'clientes-v2-tratados.csv' em um DataFrame pandas chamado 'df'.

print(df.head()) # Exibe as primeiras 5 linhas do DataFrame para uma visualização inicial dos dados.

# Codificação one-hot para 'estado_civil'
# Usa pd.get_dummies para criar novas colunas binárias para cada categoria única na coluna 'estado_civil'.
# O prefixo 'estado_civil' é adicionado aos nomes das novas colunas para identificação.
# Em seguida, as novas colunas são concatenadas com o DataFrame original usando pd.concat, e a coluna original 'estado_civil' é removida implicitamente na próxima reatribuição a df.
df = pd.concat([df, pd.get_dummies(df['estado_civil'], prefix='estado_civil')], axis=1)

print("\nDataFrame após codificação one-hot para 'estado_civil': \n", df.head()) # Exibe o DataFrame após a codificação one-hot de 'estado_civil'.

#Codificação ordinal para 'nivel_educação'
# Define um dicionário para mapear os níveis de educação a valores numéricos ordinais.
educacao_ordem = {'Ensino Fundamental': 1, 'Ensino Médio': 2, 'Ensino superior': 3, 'Pós-graduação': 4}
# Cria uma nova coluna 'nivel_educacao_ordinal' aplicando o mapeamento definido à coluna 'nivel_educacao'.
df['nivel_educacao_ordinal'] = df['nivel_educacao'].map(educacao_ordem)

print("\nDataFrame após codificação ordinal para 'nivel_educação': \n", df.head()) # Exibe o DataFrame após a codificação ordinal de 'nivel_educação'.

# Transformar 'area_atuacao' em categorias codificadas usando o método .cat.codes
# Converte a coluna 'area_atuacao' para o tipo categórico e então usa .cat.codes para atribuir um código numérico a cada categoria única.
df['area_atuacao_cod'] = df['area_atuacao'].astype('category').cat.codes

print("\nDataFrame após transformar 'area_atuacao' em códigos numéricos: \n", df.head()) # Exibe o DataFrame após a codificação categórica de 'area_atuacao'.

# LabelEncoder para 'estado'
# Label Encoder Converte cada valor único em números de 0 a n_classes-1
label_encoder = LabelEncoder() # Instancia um objeto LabelEncoder.
# Aplica o LabelEncoder à coluna 'estado', transformando seus valores categóricos em números inteiros (0, 1, 2, ...).
df['estado_code'] = label_encoder.fit_transform(df['estado'])

print("\nDataFrame após aplicar LabelEncoder em 'estado': \n", df.head()) # Exibe o DataFrame após a aplicação do LabelEncoder em 'estado'.
```

### Explicação de Termos Comuns em Manipulação de Dados

Ao trabalhar com bibliotecas como `pandas` e `scikit-learn`, alguns termos são frequentemente utilizados e é útil compreender seus significados:

*   **`axis` (eixo)**: Em estruturas de dados tabulares (como DataFrames), `axis` refere-se à dimensão ao longo da qual uma operação é realizada.
    *   `axis=0` geralmente indica operações por linha (ao longo do índice).
    *   `axis=1` geralmente indica operações por coluna.
    
    Por exemplo, ao remover colunas, você especifica `axis=1`; ao remover linhas, `axis=0`.

*   **`None` (Nulo)**: É um objeto que representa a ausência de um valor ou um valor nulo. Em Python, é um tipo de dado singleton. Em `pandas`, ele pode aparecer em dados como `NaN` (Not a Number) para valores numéricos ou como `None` para outros tipos, indicando dados ausentes.

*   **`errors` (erros)**: Este parâmetro é comum em funções que lidam com conversão de tipos de dados ou outras operações que podem falhar. Ele define como a função deve se comportar quando encontra um erro.
    *   **`'raise'` (padrão)**: Levanta uma exceção (erro) se encontrar um problema.
    *   **`'coerce'`**: Converte valores inválidos em um valor nulo (como `NaN` no pandas), permitindo que a operação continue sem interromper o programa.
    *   **`'ignore'`**: Ignora os erros e retorna a entrada como está, sem modificar os valores inválidos (pode não ser aplicável a todas as funções).

*   **`inplace` (no local)**: Um parâmetro booleano (True/False) comum em muitos métodos de `pandas`. Quando `True`, a operação é realizada diretamente no DataFrame original (modificando-o 'no local') e o método retorna `None`. Quando `False` (o padrão para muitos métodos), a operação retorna uma *nova* cópia do DataFrame com as modificações, deixando o DataFrame original inalterado.

*   **`index` (índice)**: O rótulo ou identificador para cada linha em um DataFrame ou cada elemento em uma `Series` do pandas. Ele fornece uma forma de acessar e alinhar dados. Pode ser numérico (como 0, 1, 2...) ou ter rótulos personalizados (datas, IDs, etc.).
