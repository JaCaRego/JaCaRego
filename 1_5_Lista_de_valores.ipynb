#[Lista]
* Pode utilizar qualque tipo de conteudo - str, int, float, etc
* Lista tem posições, sempre iniciando da posição zero
* Os valores podem ser modificados(adicionar dados)


```python
lista = [5, 'Jales', 8,9]
type(lista)
```

```python
lista = [5, 'Jales', 8.9]
lista.append("Carvalho") #adicionar um elemento na lista
lista.append(10)

# O erro ocorre porque a lista contém tipos de dados mistos (strings e números).
# O método sort() não sabe como comparar uma string com um número, resultando em um TypeError.
# Para corrigir, filtramos a lista para incluir apenas os elementos numéricos antes de ordenar.
lista = [item for item in lista if isinstance(item, (int, float))]

lista.sort() # Agora a ordenação funcionará
lista.reverse() #inverter a lista
lista.index(10)
# Correção: O método insert() espera (indice, valor), com o índice sendo um inteiro.
# Alterado de 'lista.insert(1.5; 2)' para 'lista.insert(1, 2)'
lista.insert(1, 2)
print(lista.index(10))
print(lista[1]) #acessar um elemento da lista, sempre começa na posição 0
print(lista)
```

#(Tupla)



* Pode utilizar qualque tipo de conteudo - str, int, float, etc
* Lista tem posições, sempre iniciando da posição zero
* Os valores (NÃO) podem ser modificados(adicionar dados), (É CONSTANTE)


```python
tupla = (5, 'Jales', 8.9)
tupla.index(8.9)
print(tupla.index(8.9))
print(tupla.count(5)) #contar quantas vezes um elemento aparece na lista
tupla.count(5) #contar quantas vezes um elemento aparece na lista
tupla[1] #acessar um elemento da lista, sempre começa na posição 0
print(tupla[1])
print(tupla)
type(tupla)
```

```python
tupla.append("Carvalho") #tuplas não suportam adição de elementos, utilizados quando tiver um valor constante.
tupla.index(8.9)
tupla[1] #acessar um elemento da lista, sempre começa na posição 0
print(tupla)
```

```python
print(tupla) # As informaçoes não podem ser alteradas (constantes)
```

#{Dicionário}
*Parecido com um [lista], mas pode utilizar qualquer valor nesse índice.
* Chave-valor, tem-se uma chave, e associa-se valor a essa chave. Ex.: Chave/índice{nome}: Valor{Jales}
* Nesse caso, não existe posições como na [Lista] e na (Tupla). Existe índice.

```python
dicionario = {'nome': 'Jales', 'cidade': 'Riachão-MA'}
dicionario.popitem
print(dicionario.popitem())
dicionario.values()
print(dicionario.values())
dicionario.keys()
print(dicionario.keys())
dicionario.items()
print(dicionario.items())
type(dicionario)
```

```python
print(dicionario)
```

```python
print(dicionario['nome']) #acessar um elemento do dicionário
```

```python
dicionario['ultimo_nome'] = 'Porto Rico' # Adicionar uma informação
```

```python
print(dicionario)
```
