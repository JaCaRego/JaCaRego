### Carregando o arquivo `ecommerce_preparados.csv`

```python
import pandas as pd

# Define o caminho para o seu arquivo CSV baixado 'ecommerce_preparados.csv'
# Certifique-se de que o arquivo foi carregado para o Colab ou especifique o caminho correto.
ecommerce_file_path = '/content/sample_data/ecommerce_preparados.csv' # Atualizado o caminho!

# Carrega o dataset
df_ecommerce = pd.read_csv(ecommerce_file_path)
print(f'Dataset "ecommerce_preparados.csv" carregado com sucesso de: {ecommerce_file_path}')

# Exibe as primeiras 5 linhas do novo DataFrame
display(df_ecommerce.head())
```

```python
import numpy as np

df_missing = df_ecommerce.copy()

# Introduzindo NaNs em algumas colunas para demonstração em df_ecommerce
df_missing.loc[df_missing.sample(frac=0.01).index, 'Nota'] = np.nan
df_missing.loc[df_missing.sample(frac=0.005).index, 'N_Avaliações'] = np.nan

print("DataFrame com NaNs simulados (df_ecommerce):")
display(df_missing.head())
```

### Identificando Valores Nulos
Podemos usar `.isnull()` em conjunto com `.sum()` para ver a contagem de valores nulos em cada coluna.

```python
print("Contagem de valores nulos por coluna:")
display(df_missing.isnull().sum())
```

### Estratégias para Corrigir Dados Nulos com `fillna()`
O método `fillna()` do pandas é muito versátil. Aqui estão algumas das estratégias mais comuns:

```python
# 1. Preencher com a média (para colunas numéricas)
df_filled_mean = df_missing.copy()
mean_nota = df_filled_mean['Nota'].mean()
df_filled_mean['Nota'] = df_filled_mean['Nota'].fillna(mean_nota)

print(f"'Nota' preenchido com a média ({mean_nota:.2f}):")
display(df_filled_mean[['Nota']].head(10))
display(df_filled_mean.isnull().sum())
```

```python
# 2. Preencher com a mediana (menos sensível a outliers)
df_filled_median = df_missing.copy()
median_n_avaliacoes = df_filled_median['N_Avaliações'].median()
df_filled_median['N_Avaliações'] = df_filled_median['N_Avaliações'].fillna(median_n_avaliacoes)

print(f"'N_Avaliações' preenchido com a mediana ({median_n_avaliacoes:.0f}):")
display(df_filled_median[['N_Avaliações']].head(10))
display(df_filled_median.isnull().sum())
```

```python
# 3. Preencher com um valor específico (por exemplo, 0 para 'Nota')
df_filled_value = df_missing.copy()
df_filled_value['Nota'] = df_filled_value['Nota'].fillna(0)

print("'Nota' preenchido com 0:")
display(df_filled_value[['Nota']].head(10))
display(df_filled_value.isnull().sum())
```

### Outras Estratégias (`ffill`, `bfill`)

`ffill` (forward fill) preenche os valores nulos com o valor da linha anterior.
`bfill` (backward fill) preenche os valores nulos com o valor da próxima linha.

Essas estratégias são mais úteis para dados de séries temporais ou quando a ordem dos dados é significativa.

```python
df_ffill = df_missing.copy()
df_ffill['Nota'] = df_ffill['Nota'].ffill()

print("'Nota' preenchido com ffill:")
display(df_ffill[['Nota']].head(10))
display(df_ffill.isnull().sum())
```

### Removendo Linhas com Valores Nulos (`dropna()`)
Em alguns casos, se a quantidade de dados nulos for pequena ou se a linha inteira não tiver utilidade sem aquele dado, podemos simplesmente remover as linhas que contêm valores nulos.

```python
df_dropped = df_missing.copy()

# Remove todas as linhas que contêm pelo menos um valor nulo
df_dropped.dropna(inplace=True)

print(f"Shape do DataFrame original com NaNs simulados: {df_missing.shape}")
print(f"Shape do DataFrame após remover linhas com NaNs: {df_dropped.shape}")
display(df_dropped.isnull().sum())
```

### Salvando o DataFrame Tratado

Depois de remover as linhas com valores nulos, é uma boa prática salvar o DataFrame resultante para uso futuro ou para outras etapas de análise, garantindo que você não precise refazer o pré-processamento. Usaremos o método `to_csv()` para isso.

```python
# Salva o DataFrame df_dropped para um novo arquivo CSV
output_file_path = '/content/df_ecommerce_preparados_corrigido.csv'
df_dropped.to_csv(output_file_path, index=False) # index=False evita escrever o índice do DataFrame como uma coluna no CSV

print(f"DataFrame salvo com sucesso em: {output_file_path}")
```

A escolha da melhor estratégia depende do contexto dos seus dados e do tipo de análise que você pretende fazer. É crucial entender o porquê dos dados estarem faltando antes de decidir como preenchê-los ou removê-los.
