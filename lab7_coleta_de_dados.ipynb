```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
requests.packages.urllib3.disable_warnings()

url = 'https://books.toscrape.com/'
requisicao = requests.get(url)

extracao = BeautifulSoup(requisicao.text, 'html.parser')

# Verificar se a requisição foi bem-sucedida
if requisicao.status_code == 200:
    # Obter o conteúdo da página
    conteudo_html = requisicao.text
    # Criar um objeto Beautiful Soup
    soup = BeautifulSoup(conteudo_html, 'html.parser')
    # Usar prettify para formatar o HTML e exibir os primeiros 2000 caracteres
    print(soup.prettify()[:2000])
else:
    print(f"Erro ao acessar o site. Status code: {requisicao.status_code}")

# Exibir tags Alinhadas
for titulo in extracao.find_all('h2'):
    print('\n Título: ', titulo.text.strip())
    for link in titulo.find_next_siblings('p'):
        for a in link.find_all('a', href=True):
            print('Texto link: ', a.text.strip(), ' | URL:', a["href"] )

```

### Exemplo 1: Extraindo dados de uma tabela

Vamos extrair dados de uma tabela simples em uma página web.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.w3schools.com/html/html_tables.asp'
requisicao = requests.get(url)
extracao = BeautifulSoup(requisicao.text, 'html.parser')

table = extracao.find('table')

if table:
    headers = [th.text for th in table.find_all('th')]
    data = []
    for row in table.find_all('tr')[1:]: # Ignorar a linha do cabeçalho
        cells = [td.text.strip() for td in row.find_all('td')]
        data.append(cells)

    df = pd.DataFrame(data, columns=headers)
    display(df)
else:
    print("Nenhuma tabela encontrada na página.")
```

### Exemplo 2: Extraindo links de uma página

Vamos extrair todos os links (`<a>` tags) de uma página web.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://books.toscrape.com/'
requisicao = requests.get(url)
extracao = BeautifulSoup(requisicao.text, 'html.parser')

for link in extracao.find_all('a', href=True):
    print(f"Texto do link: {link.text.strip()} | URL: {link['href']}")
```

### Exemplo 3: Extraindo informações específicas de elementos com classes CSS

Vamos extrair títulos de livros e preços de uma página com base nas classes CSS.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://books.toscrape.com/'
requisicao = requests.get(url)
extracao = BeautifulSoup(requisicao.text, 'html.parser')

for article in extracao.find_all('article', class_='product_pod'):
    title = article.find('h3').find('a')['title']
    price = article.find('p', class_='price_color').text.strip()
    print(f"Título: {title} | Preço: {price}")
```

5.  **Cria um objeto BeautifulSoup:** Cria um objeto `BeautifulSoup` a partir do conteúdo HTML da resposta, usando o parser `html.parser`.

7.  **Exibe o HTML formatado:** Se a requisição foi bem-sucedida, formata o HTML usando `prettify()` e exibe os primeiros 2000 caracteres.
8.  **Exibe títulos e links:** Itera sobre todas as tags `<h2>` encontradas no HTML e imprime o texto do título. Em seguida, para cada título, encontra as tags `p` que são irmãs subsequentes e, dentro delas, encontra as tags `<a>` (links) com o atributo `href`. Para cada link encontrado, imprime o texto do link e a URL.

Em resumo, o código acessa um site de livros, verifica se a página foi carregada corretamente, imprime uma parte do código HTML da página e, em seguida, extrai e imprime os títulos dos livros e os links associados a eles.
