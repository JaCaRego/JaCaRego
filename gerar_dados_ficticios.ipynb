```python
import pandas as pd
import random
from faker import Faker
'''
Pandas: É uma biblioteca usada para manipulação e análise de dados. Ele fornece
estruturas de dados como DataFrames que facilitam o trabalho com dados tabulares
(como planilhas ou tabelas de banco de dados).
Random: É uma biblioteca para gerar números pseudoaleatórios. É útil para
simulações, jogos ou qualquer situação em que você precise de aleatoriedade.
Faker: É uma biblioteca para gerar dados falsos (mas realistas). É ótimo para
criar dados de teste para seus aplicativos, como nomes, endereços, números de
telefone, etc.

'''

faker = Faker('pt_BR')

dados_pessoas = []

'''
A expressão for _ in range(10): em Python é usada para criar um loop que executa
um bloco de código um número específico de vezes. range(10) gera uma sequência
de números de 0 a 9 (total de 10 números).
O _ (underscore) é uma convenção em Python para indicar que a variável do
loop não será usada dentro do bloco de código. Neste caso, você só precisa
que o loop execute 10 vezes, independentemente do valor do índice.
Então, for _ in range(10): simplesmente executa o código dentro do loop 10 vezes.

'''

for _ in range(10):
    name = faker.name()
    cpf = faker.cpf()
    idade = random.randint(18, 60)
    data = faker.date_of_birth(minimum_age=idade, maximum_age=idade).strftime('%d/%m/%Y')
    endereco = faker.address()
    estado = faker.state()
    pais = 'Brasil'

    pessoa = {
        'nome': name,
        'cpf': cpf,
        'idade': idade,
        'data': data,
        'endereco': endereco,
        'estado': estado,
        'pais': pais,
    }

    dados_pessoas.append(pessoa)

'''
O .append() é um método comum em Python, usado principalmente com listas.
No código que você compartilhou, a linha dados_pessoas.append(pessoa) significa
que o dicionário pessoa, que contém os dados gerados para uma única pessoa,
está sendo adicionado ao final da lista dados_pessoas.

Então, a cada vez que o loop for _ in range(10): executa, um novo dicionário
com dados de uma pessoa é criado e adicionado à lista dados_pessoas. Ao final
do loop, a lista dados_pessoas terá 10 dicionários, cada um representando
uma pessoa diferente.

'''

df_pessoas = pd.DataFrame(dados_pessoas)
print(df_pessoas)

# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_colwidth', None)
# pd.set_option('display.width', None)

print(df_pessoas.to_string()) # head() tail()

df_pessoas.to_csv('pessoas.csv')
```
