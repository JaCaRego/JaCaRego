```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Exemplo: Criação de um gráfico de dispersão com Matplotlib e Seaborn

# Dados de exemplo
data = {
    'idade': [23, 45, 56, 25, 34, 42, 67, 29, 38, 50],
    'salario': [50000, 80000, 120000, 45000, 60000, 75000, 130000, 48000, 70000, 90000]
}
df = pd.DataFrame(data)

# Gráfico de dispersão com Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(df['idade'], df['salario'], color='blue', alpha=0.5)
plt.title('Relação entre Idade e Salário')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.show()

# Gráfico de dispersão com Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='idade', y='salario', data=df, hue='idade', palette='viridis')
plt.title('Relação entre Idade e Salário')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.show()

```

## Importância da Exploração de Dados (EDA) antes da Visualização

A Exploração de Dados (EDA - Exploratory Data Analysis) é uma etapa fundamental e muitas vezes subestimada no processo de análise de dados. Ela consiste em examinar os dados disponíveis para entender suas características principais, identificar padrões, detectar anomalias, testar hipóteses e validar suposições, geralmente com a ajuda de métodos estatísticos e representações gráficas.

### Por que a EDA é Crucial Antes da Visualização?

1.  **Entendimento Profundo dos Dados:** Permite conhecer a estrutura dos dados, os tipos de variáveis (numéricas, categóricas, textuais), seus formatos e o significado de cada coluna. Isso evita interpretações erradas na fase de visualização.
2.  **Qualidade dos Dados:** Ajuda a identificar e lidar com problemas como valores ausentes (NaN), outliers (pontos fora do padrão), inconsistências e erros de digitação. Visualizar dados "sujos" pode levar a gráficos enganosos e conclusões erradas.
3.  **Identificação de Padrões e Relações:** A EDA revela correlações entre variáveis, distribuições de dados e tendências, que podem guiar a escolha dos gráficos mais adequados para destacar essas descobertas.
4.  **Formulação de Hipóteses:** Ao explorar os dados, novas perguntas e hipóteses podem surgir, direcionando futuras análises e decisões de modelagem.
5.  **Otimização da Visualização:** Com um bom entendimento dos dados, você pode selecionar os tipos de gráficos mais eficazes, escolher as escalas corretas e focar nos aspectos mais relevantes para a sua audiência, tornando a visualização mais impactante e precisa.
6.  **Evitar "Garbage In, Garbage Out":** Visualizar dados sem antes explorá-los é como construir uma casa em um terreno sem inspecioná-lo; o resultado pode ser instável e problemático.

### Principais Técnicas de EDA:

*   **Estatísticas Descritivas:** Calcular média, mediana, moda, desvio padrão, quartis, amplitude, etc., para ter um panorama numérico das variáveis.
*   **Verificação de Valores Ausentes:** Identificar a quantidade e o padrão de dados faltantes e decidir sobre a estratégia de tratamento (remover, imputar, etc.).
*   **Detecção de Outliers:** Identificar valores extremos que podem distorcer análises e gráficos.
*   **Análise de Distribuição:** Usar histogramas, gráficos de densidade (KDE plots) e box plots para entender como os dados estão distribuídos.
*   **Análise de Correlação:** Investigar a relação entre duas ou mais variáveis usando gráficos de dispersão (scatter plots) e mapas de calor (heatmaps) para matrizes de correlação.
*   **Agrupamento e Segmentação:** Dividir os dados em grupos para comparar características e padrões.
*   **Engenharia de Features:** Criar novas variáveis a partir das existentes para extrair mais informações.

### Bibliotecas Essenciais para EDA em Python:

*   **`Pandas`:** A biblioteca fundamental para manipulação e análise de dados. Essencial para carregar, limpar, transformar e resumir DataFrames.
*   **`NumPy`:** Base para computação numérica em Python, frequentemente usada em conjunto com Pandas para operações matemáticas eficientes.
*   **`Matplotlib`:** A biblioteca de visualização de baixo nível, que permite um controle granular sobre cada elemento do gráfico. É a base para muitas outras bibliotecas de visualização.
*   **`Seaborn`:** Construída sobre Matplotlib, oferece uma interface de alto nível para criar gráficos estatísticos atraentes e informativos com menos código, sendo excelente para distribuições, relações e comparações.
*   **`Plotly` / `Bokeh`:** Para visualizações interativas que podem ser muito úteis na fase de exploração, permitindo zoom, hover e filtros dinâmicos.

Realizar uma EDA robusta garante que as visualizações subsequentes sejam não apenas esteticamente agradáveis, mas, mais importante, precisas, informativas e capazes de comunicar insights valiosos de forma clara.

### Como Escolher o Tipo de Gráfico Ideal?

Para escolher o gráfico adequado, considere primeiramente **o tipo de dado** que você possui e **o objetivo** da sua visualização. Cada gráfico é otimizado para um propósito específico.

**1. Para Mostrar Composição (Partes de um Todo):**
   *   **Objetivo:** Entender a proporção de cada categoria em relação ao total.
   *   **Gráficos Recomendados:**
       *   **Gráfico de Pizza (Pie Chart):** Bom para 2-5 categorias, onde a soma total é 100%. Evite para muitas categorias, pois fica difícil comparar.
       *   **Gráfico de Barras Empilhadas (Stacked Bar Chart):** Excelente para comparar a composição entre diferentes categorias ou ao longo do tempo.

**2. Para Mostrar Distribuição (Frequência ou Concentração):**
   *   **Objetivo:** Analisar a forma como os dados estão distribuídos, identificar tendências centrais, dispersão e outliers.
   *   **Gráficos Recomendados:**
       *   **Histograma:** Ideal para variáveis numéricas contínuas, mostra a frequência de dados em intervalos (bins).
       *   **Gráfico de Densidade (KDE Plot):** Similar ao histograma, mas com uma curva suave, ótimo para identificar modas e comparar distribuições.
       *   **Box Plot (Diagrama de Caixa):** Mostra a mediana, quartis e outliers de uma distribuição, excelente para comparar distribuições entre grupos.

**3. Para Mostrar Relação (Entre Variáveis):**
   *   **Objetivo:** Entender a conexão ou correlação entre duas ou mais variáveis.
   *   **Gráficos Recomendados:**
       *   **Gráfico de Dispersão (Scatter Plot):** Perfeito para duas variáveis numéricas, revela padrões, correlações (positiva, negativa, nula) e agrupamentos.
       *   **Gráfico de Bolhas (Bubble Chart):** Uma variação do scatter plot que adiciona uma terceira variável representada pelo tamanho da bolha.
       *   **Mapa de Calor (Heatmap):** Útil para visualizar a correlação entre muitas variáveis em uma matriz.

**4. Para Mostrar Comparação (Entre Categorias ou ao Longo do Tempo):**
   *   **Objetivo:** Comparar valores entre diferentes itens ou observar mudanças ao longo do tempo.
   *   **Gráficos Recomendados:**
       *   **Gráfico de Barras (Bar Chart):** Muito versátil para comparar quantidades entre categorias discretas (quantitativas ou ordinais).
       *   **Gráfico de Linhas (Line Chart):** O melhor para mostrar tendências e mudanças de uma ou mais variáveis numéricas ao longo do tempo (séries temporais).
       *   **Gráfico de Colunas (Column Chart):** Similar ao gráfico de barras, mas com barras verticais, usado frequentemente para comparar valores em categorias ou períodos.

**5. Para Mostrar Localização Geográfica:**
   *   **Objetivo:** Visualizar dados em um contexto geográfico.
   *   **Gráficos Recomendados:**
       *   **Mapas de Coroplético (Choropleth Maps):** Usam sombreamento ou padrões para representar dados estatísticos por regiões geográficas.
       *   **Mapas de Pontos (Point Maps):** Representam eventos ou ocorrências em locais específicos.

## Gráficos de Densidade (KDE - Kernel Density Estimation)

Gráficos de densidade, frequentemente chamados de *KDE plots* (Kernel Density Estimation), são visualizações poderosas para entender a distribuição de uma variável numérica. Em vez de barras discretas como nos histogramas, os gráficos de densidade usam uma função contínua para estimar a Função Densidade de Probabilidade (PDF - uma função que descreve a probabilidade de uma variável aleatória assumir um determinado valor) da variável. Isso resulta em uma curva suave que mostra onde os valores estão mais concentrados e onde são mais raros.

### Funções e Utilidade:

1.  **Visualização da Distribuição Contínua**: Oferecem uma representação suave da distribuição dos dados, o que é especialmente útil para variáveis contínuas, onde a granularidade dos *bins* (intervalos ou "caixas" que agrupam dados em um histograma) de um histograma pode ocultar padrões.
2.  **Identificação de Picos (Modas)**: Permitem identificar facilmente as modas (picos) da distribuição, indicando os valores mais frequentes na variável.
3.  **Comparação de Distribuições**: São excelentes para comparar a distribuição de uma variável entre diferentes grupos ou categorias, sobrepondo múltiplas curvas de densidade no mesmo gráfico.
4.  **Detecção de Assimetria e Curtose**: Ajuda a visualizar se a distribuição é simétrica, assimétrica (skewed), ou se possui caudas pesadas (curtose - medida do "achatamento" ou "alongamento" das caudas de uma distribuição).
5.  **Substituição para Histogramas**: Em muitos casos, são preferíveis aos histogramas quando se deseja uma representação mais limpa e menos dependente da escolha do tamanho do *bin* (intervalo de agrupamento).

*   **`Seaborn`**: É a biblioteca mais indicada e popular para criar gráficos de densidade no Python.     *   **Melhor Utilidade**: Ideal para análises exploratórias rápidas, visualização de distribuições univariadas (de uma única variável) e bivariadas (de duas variáveis), e comparação de distribuições entre grupos com facilidade.
