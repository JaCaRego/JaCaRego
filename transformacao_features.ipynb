# Engenharia de features

O principal objetivo deste código é realizar uma série de transformações em um DataFrame pandas (df). Ele executa as seguintes operações:

*   **Leitura de dados**: Carrega um arquivo CSV (`dados/clientes-v2-tratados.csv`) em um DataFrame.
*   **Transformação Logarítmica**: Aplica a transformação logarítmica (usando `np.log1p`) à coluna 'salario' para criar 'salario_log'. Isso é útil para normalizar dados com distribuição assimétrica.
*   **Transformação Box-Cox**: Aplica a transformação Box-Cox à coluna 'salario' (após adicionar 1 para lidar com zeros) para criar 'salario_boxcox'. Esta é outra técnica de transformação de potência para tornar os dados mais parecidos com uma distribuição normal.
*   **Codificação de Frequência**: Calcula a frequência de cada categoria na coluna 'estado' e usa-a para criar uma nova coluna 'estado_freq', substituindo os valores categóricos por suas frequências relativas. Isso pode ser útil para modelos que não lidam bem com variáveis categóricas diretas.
*   **Interações de Variáveis**: Cria uma nova coluna 'interacao_idade_filhos' multiplicando os valores das colunas 'idade' e 'numero_filhos'. Isso pode capturar relações não-lineares entre as variáveis.

```python
import pandas as pd  # Importa a biblioteca pandas, essencial para manipulação e análise de dados em formato de tabela (DataFrames).
import numpy as np   # Importa a biblioteca numpy, utilizada para operações numéricas e matemáticas, especialmente com arrays.
from scipy import stats # Importa o módulo stats da biblioteca SciPy, que contém funções para estatística, incluindo transformações como Box-Cox.

pd.set_option('display.width', None) # Configura a opção de exibição do pandas para que DataFrames não sejam truncados na largura, mostrando todas as colunas.

df = pd.read_csv('dados/clientes-v2-tratados.csv') # Lê um arquivo CSV localizado no caminho especificado e o carrega em um DataFrame pandas chamado 'df'.

print(df.head()) # Imprime as primeiras 5 linhas do DataFrame 'df', permitindo uma visualização rápida da estrutura e dos dados.

# Transformação Logarítmica
df['salario_log'] = np.log1p(df['salario']) # Aplica a transformação logarítmica (log(1+x)) à coluna 'salario' e armazena o resultado em uma nova coluna 'salario_log'.
                                           # np.log1p é usado para lidar com valores de 'salario' que podem ser zero, evitando erros.

print("\nDataFrame após transformação logarítmica no 'salario': \n", df.head()) # Imprime as primeiras linhas do DataFrame após a criação da coluna 'salario_log'.

# Transformação Box-Cox
df['salario_boxcox'], _ = stats.boxcox(df['salario'] + 1) # Aplica a transformação Box-Cox à coluna 'salario' (adicionando 1 para garantir valores positivos, requisito da transformação)
                                                      # O resultado é armazenado na nova coluna 'salario_boxcox'. O segundo valor retornado pela função (lambda) é descartado com '_'.

print("\nDataFrame após transformação Box_Cox no 'salario': \n", df.head()) # Imprime as primeiras linhas do DataFrame após a criação da coluna 'salario_boxcox'.

# Codificação de Frequência para 'estado'
estado_freq = df['estado'].value_counts() / len(df) # Calcula a frequência relativa de cada categoria na coluna 'estado'.
                                                 # 'value_counts()' conta as ocorrências e a divisão por 'len(df)' normaliza para frequências.
df['estado_freq'] = df['estado'].map(estado_freq) # Mapeia as categorias da coluna 'estado' para suas respectivas frequências relativas, criando a nova coluna 'estado_freq'.

print("\nDataFrame após codificação de frequência para 'estado': \n", df.head()) # Imprime as primeiras linhas do DataFrame após a criação da coluna 'estado_freq'.

# Interações
df['interacao_idade_filhos'] = df['idade'] * df['numero_filhos'] # Cria uma nova coluna 'interacao_idade_filhos' multiplicando os valores das colunas 'idade' e 'numero_filhos'.
                                                               # Isso pode capturar relações não-lineares ou sinérgicas entre as variáveis.
print("\nDataFrame após criação de interações entre 'idade' e 'numero_filhos': \n", df.head()) # Imprime as primeiras linhas do DataFrame após a criação da coluna 'interacao_idade_filhos'.
```

## Explicação do termo 'lambda' no seu código

No seu código, especificamente na linha:

`df['salario_boxcox'], _ = stats.boxcox(df['salario'] + 1)`

a palavra `_` (underscore) está recebendo o valor do **parâmetro \( \lambda \)** (lê-se 'lambda') da transformação Box-Cox, e não se refere a uma função `lambda` do Python.

### Detalhes:
1.  **`stats.boxcox(df['salario'] + 1)`**: Esta função da biblioteca SciPy aplica a transformação Box-Cox. Ela sempre retorna dois valores:
    *   Os dados transformados (que você atribui a `df['salario_boxcox']`).
    *   O valor ótimo do parâmetro \( \lambda \) que foi encontrado e usado na transformação para tornar os dados mais próximos de uma distribuição normal.
2.  **`_` (underscore)**: Em Python, é uma convenção comum usar `_` para atribuir um valor que não será utilizado posteriormente no código. Neste caso, o valor do parâmetro \( \lambda \) da transformação Box-Cox é capturado por `_`, indicando que você não pretende usá-lo ou visualizá-lo diretamente, mas ele é retornado pela função `stats.boxcox`.

## Coleção de Itens: Detalhes e Exemplos

Quando falamos em **'coleção de itens'**, estamos nos referindo a uma estrutura de dados que agrupa múltiplos valores sob um único nome ou identificador. A forma como esses itens são armazenados e manipulados varia de acordo com o tipo de coleção.

### 1. Listas (Python built-in)

São coleções **ordenadas** e **mutáveis**, ou seja, você pode adicionar, remover ou modificar itens após a criação. Podem conter itens de diferentes tipos.

**Exemplo:**

```python
minha_lista = [1, 'dois', 3.0, True] # Coleção de itens de tipos variados
print(f"Lista original: {minha_lista}")

minha_lista.append(5) # Adiciona um item
print(f"Lista após adicionar: {minha_lista}")

minha_lista[0] = 10 # Modifica um item
print(f"Lista após modificar: {minha_lista}")
```

### 2. Tuplas (Python built-in)

São coleções **ordenadas** e **imutáveis**. Uma vez criada, uma tupla não pode ser alterada (não é possível adicionar, remover ou modificar itens). Também podem conter itens de diferentes tipos.

**Exemplo:**

```python
minha_tupla = (1, 'dois', 3.0, False) # Coleção de itens de tipos variados
print(f"Tupla original: {minha_tupla}")

# Tentativa de modificar um item resultaria em erro:
# minha_tupla[0] = 10
# TypeError: 'tuple' object does not support item assignment

# Tentativa de adicionar um item resultaria em erro:
# minha_tupla.append(5)
# AttributeError: 'tuple' object has no attribute 'append'
```

### 3. Arrays NumPy

São coleções **ordenadas** e **mutáveis**, mas com uma característica fundamental: são otimizados para armazenar **itens do mesmo tipo (homogêneos)**. Isso os torna extremamente eficientes para cálculos numéricos e manipulação de grandes volumes de dados. São a base para muitas operações de ciência de dados e aprendizado de máquina.

**Exemplo:**

```python
import numpy as np

meu_array = np.array([1, 2, 3, 4, 5]) # Coleção de itens do mesmo tipo (inteiros)
print(f"Array original: {meu_array}")
print(f"Tipo dos elementos do array: {meu_array.dtype}")

meu_array[0] = 10 # Modifica um item
print(f"Array após modificar: {meu_array}")

# Arrays NumPy permitem operações vetorizadas eficientes
meu_array_multiplicado = meu_array * 2
print(f"Array multiplicado por 2: {meu_array_multiplicado}")

# Se você tentar adicionar um tipo diferente, o NumPy pode converter (upcast) todos os elementos
array_misto = np.array([1, 'dois', 3.0])
print(f"Array misto (notar a conversão para string): {array_misto}")
print(f"Tipo dos elementos do array misto: {array_misto.dtype}") # Todos se tornam strings
```

Em resumo, quando menciono 'coleção de itens', estou me referindo a qualquer uma dessas estruturas que agrupam dados. A escolha entre elas depende das suas necessidades: se precisa de mutabilidade, imutabilidade, heterogeneidade de tipos, ou a performance e capacidades numéricas dos arrays NumPy.

Quando nos referimos a "coleção de itens" no contexto de arrays, estamos falando de **dados individuais** que são armazenados juntos. Pense como uma lista ou uma tupla em Python.

*   **Listas e Tuplas** em Python são coleções de itens. A diferença é que listas são mutáveis (podem ser alteradas) e tuplas são imutáveis (não podem ser alteradas).
*   Um **array NumPy** é uma estrutura que armazena uma coleção de itens (geralmente do mesmo tipo) de forma muito eficiente para operações numéricas.
*   Uma **biblioteca** (como NumPy ou Pandas) é um conjunto de módulos e funções que você pode usar em seu código; ela não é uma 'coleção de itens' no sentido de dados que estão sendo armazenados em uma estrutura.

## O que são Arrays?

Arrays são estruturas de dados que armazenam uma coleção de itens do mesmo tipo (homogêneos) em uma sequência contígua de memória. São otimizados para operações numéricas rápidas e acesso eficiente aos elementos por meio de índices. Em Python, a biblioteca **NumPy** é amplamente utilizada para criar e manipular arrays (`ndarray`) multidimensionais.

```python
import pandas as pd
import numpy as np
from scipy import stats

pd.set_option('display.width', None)

df = pd.read_csv('dados/clientes-v2-tratados.csv')

print(df.head())

# Transformação Logarítmica
df['salario_log'] = np.log1p(df['salario']) # log1p é usado para evitar problemas com valores zero

print("\nDataFrame após transformação logarítmica no 'salario': \n", df.head())

# Transformação Box-Cox
df['salario_boxcox'], _ = stats.boxcox(df['salario'] + 1)

print("\nDataFrame após transformação Box_Cox no 'salario': \n", df.head())

# Codificação de Frequência para 'estado'
estado_freq = df['estado'].value_counts() / len(df)
df['estado_freq'] = df['estado'].map(estado_freq)

print("\nDataFrame após codificação de frequência para 'estado': \n", df.head())

# Interações
df['interacao_idade_filhos'] = df['idade'] * df['numero_filhos']
print("\nDataFrame após criação de interações entre 'idade' e 'numero_filhos': \n", df.head())
```
