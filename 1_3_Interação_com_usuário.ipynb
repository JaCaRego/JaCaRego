###Erros comuns de typagem

```python
desconto = valor - 5 # não consegue fazer uma opeação, num campo string
```

### Soluções para erros comuns de typagem

* Solução caso for utiçoza a operação varias vezes no codigo (criar variável)

```python
# solução caso for usar o valor varias vezes no codigo
valor_convert = float(valor)
desconto = valor_convert - 5
print('R$', desconto)
valor_convert = int(valor_convert)
desconto = valor_convert - 5
print('R$', desconto)
```

* Caso for utilizar apenas uma vez no codigo, usar:

```python
desconto = float(valor) - 5
print('R$', desconto)
desconto = int(float(valor)) - 5
print('R$', desconto)
```
