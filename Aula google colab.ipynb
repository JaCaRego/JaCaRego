4.3. Operações

As operações de variáveis do tipo string são:


*   + (concatenação).

Exemplo: Nome Completo.




```python
nome = 'Andre Marcos'
sobrenome = 'Perez'
apresentacao ='Ola, meu nome é ' + nome + ' ' + sobrenome+'.'
print(apresentacao)
```

Uma outra forma de concatenar strings é utilizar operaçoes de formatação:


```python
nome = 'Andre Marcos'
sobrenome = 'Perez'
apresentacao = f'Ola, meu nome é {nome} {sobrenome}.'
print(apresentacao)
```

Outra operação muito utilizada é a de fatiamento(slicing)

Exempro: Informações de email.



```python
email = 'andre.perez@gmail.com'
print('0: ' + email[0])
print('11: ' + email[11])

print('-1:' + email[-1])
print('-2:' + email[-2])
```
