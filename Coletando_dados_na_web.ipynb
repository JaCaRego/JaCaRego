### Resumo sobre HTML Parser, `strip()`, `find()` e Bibliotecas

*   **`requests`**: Esta biblioteca é usada para fazer requisições HTTP, ou seja, para baixar o conteúdo de uma página web a partir de uma URL. `requests.get(url)` obtém o HTML da página.

*   **`BeautifulSoup`**: Depois que o `requests` pega o conteúdo HTML, o `BeautifulSoup` (do módulo `bs4`) analisa e estrutura os dados HTML/XML. Ele transforma o texto HTML bruto em um objeto Python complexo que pode ser facilmente pesquisado.
    *   **`html.parser`**: É um dos *parsers* que o `BeautifulSoup` pode usar. Ele é o parser padrão e nativo do Python para analisar o HTML, criando a estrutura que o BeautifulSoup entende.

*   **`find_all()`** (e `find()`): São métodos do objeto `BeautifulSoup` que permitem buscar elementos HTML específicos (como tags `<h2>`, `<p>`, `<a>`) dentro da estrutura da página. Você pode procurar por nome da tag, atributos (como `href=True`), entre outros. Por exemplo, `extracao.find_all('h2')` encontra todas as tags `<h2>`.

*   **`.strip()`**: Este é um método de strings em Python. Ele é usado para remover espaços em branco (como espaços, quebras de linha e tabulações) do início e do fim de uma string. `linha_texto.text.strip()` garante que o texto extraído da tag não tenha espaços indesejados.

### Explicação do atributo `href`

O atributo `href`, abreviação de 'hypertext reference', é uma parte fundamental das tags de âncora HTML (`<a>`). Ele especifica a URL (Uniform Resource Locator) para a qual o hiperlink aponta. Quando um usuário clica em um link, o navegador navega para o endereço especificado no atributo `href`.

No contexto do código Python, `a['href']` está acessando o valor do atributo `href` para cada tag de âncora (`<a>`) encontrada na página da web. Isso permite que o código extraia a URL de destino de cada link.

### Como usar `href=True` no seu código para extrair links

No seu código, na célula `4KhPSK18IVR-`, você já possui a lógica pronta para extrair links usando `href=True`, mas ela está comentada. Para ativá-la, basta descomentar as linhas relevantes.

Veja a seção do seu código que você pode descomentar:

```python
# # Exibir tags Alinhadas
for titulo in extracao.find_all('h2'):
    print('\n Título: ', titulo.text.strip())
    for link in titulo.find_next_siblings('p'):
        for a in link.find_all('a', href=True):
            print('Texto link: ', a.text.strip(), ' | URL:', a["href"] )
```

**Para fazer isso:**

1.  Vá para a célula de código `4KhPSK18IVR-`.
2.  Remova o caractere `#` do início das linhas do laço `for` que começa com `for titulo in extracao.find_all('h2'):` e todas as linhas seguintes dentro desse laço.
3.  Após descomentar, execute a célula novamente. O código então buscará todas as tags `<h2>`, e para cada uma, encontrará os parágrafos subsequentes (`<p>`) que contêm links (`<a>` com atributo `href`). Ele imprimirá o texto visível do link e a URL (`a['href']`) de cada um.
