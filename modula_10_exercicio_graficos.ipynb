```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o arquivo CSV em um DataFrame
try:
    df_new = pd.read_csv('ecommerce_preparados_final.csv')
    print('Arquivo "ecommerce_preparados_final.csv" carregado com sucesso!')
except FileNotFoundError:
    print('Erro: O arquivo "ecommerce_preparados_final.csv" não foi encontrado. Por favor, certifique-se de que o arquivo está no diretório correto.')
    df_new = None # Define df_new como None para evitar erros posteriores
```

```python
# Histograma: coluna 'Preço' está correta
# Escolha da variável: 'Preço' é uma variável numérica contínua, ideal para um histograma que mostra a distribuição dos valores e a frequência em que aparecem.
# Biblioteca e Método: O histograma é gerado usando a biblioteca `seaborn` com o método `histplot`.
column_hist = 'Preço'

plt.figure(figsize=(10, 6))
sns.histplot(df_new[column_hist], kde=True, bins=30)
plt.title(f'Distribuição de {column_hist}')
plt.xlabel(column_hist)
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Gráfico de Dispersão: 'Preço' vs 'Nota'
# Escolha das variáveis: 'Preço' e 'Nota' são ambas variáveis numéricas. O gráfico de dispersão é usado para investigar a relação ou correlação entre duas variáveis quantitativas.
# Biblioteca e Método: O gráfico de dispersão é gerado usando a biblioteca `seaborn` com o método `scatterplot`.
column_scatter_x = 'Preço'
column_scatter_y = 'Nota' # Corrigido de 'Avaliação' para 'Nota'

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_new[column_scatter_x], y=df_new[column_scatter_y])
plt.title(f'Relação entre {column_scatter_x} e {column_scatter_y}')
plt.xlabel(column_scatter_x)
plt.ylabel(column_scatter_y)
plt.grid(True, alpha=0.75)
plt.show()
```

```python
numeric_cols = df_new.select_dtypes(include=['number'])
# Escolha das variáveis: Selecionamos todas as colunas numéricas para analisar as correlações entre elas. Um mapa de calor é eficaz para visualizar a matriz de correlação de múltiplas variáveis.
# Biblioteca e Método: O mapa de calor é gerado usando a biblioteca `seaborn` com o método `heatmap`.
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Calor da Matriz de Correlação')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de Barras: Usando a coluna 'Marca'
# Escolha da variável: 'Marca' é uma variável categórica. O gráfico de barras é adequado para mostrar a contagem de ocorrências de cada categoria.
# Biblioteca e Método: O gráfico de barras é gerado usando a biblioteca `seaborn` com o método `countplot`.
column_bar = 'Marca'

# Obter as top 15 marcas para o gráfico de barras
top_n = 15
brand_counts = df_new[column_bar].value_counts()
top_brands = brand_counts.head(top_n).index

df_top_brands = df_new[df_new[column_bar].isin(top_brands)]

plt.figure(figsize=(16, 9))
sns.countplot(y=column_bar, data=df_top_brands, order=top_brands, palette='viridis', hue=column_bar, legend=False)
plt.title(f'Contagem dos {top_n} Itens Mais Frequentes por {column_bar}', fontsize=18)
plt.xlabel('Número de Itens', fontsize=14)
plt.ylabel(column_bar, fontsize=14)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Gráfico de Pizza: Usando a coluna 'Marca'
# Escolha da variável: 'Marca' é uma variável categórica. O gráfico de pizza é adequado para mostrar a proporção de cada categoria (as 5 principais marcas, neste caso) em relação ao total.
# Biblioteca e Método: O gráfico de pizza é gerado usando a biblioteca `matplotlib.pyplot` com o método `pie`.
column_pie = 'Marca'

# Obter as top 5 categorias para o gráfico de pizza
category_counts_pie = df_new[column_pie].value_counts().head(5)
plt.figure(figsize=(10, 10))
plt.pie(category_counts_pie, labels=category_counts_pie.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title(f'Proporção das 5 Maiores {column_pie}s')
plt.axis('equal')
plt.show()
```

```python
# Gráfico de Densidade: 'Nota'
# Escolha da variável: 'Nota' é uma variável numérica contínua. Um gráfico de densidade (KDE) é similar ao histograma, mas mostra a distribuição dos dados de forma mais suave, destacando a forma e picos da distribuição.
# Biblioteca e Método: O gráfico de densidade é gerado usando a biblioteca `seaborn` com o método `kdeplot`.
column_kde = 'Nota' # Corrigido de 'Avaliação' para 'Nota'

plt.figure(figsize=(10, 6))
sns.kdeplot(df_new[column_kde], fill=True)
plt.title(f'Gráfico de Densidade de {column_kde}')
plt.xlabel(column_kde)
plt.ylabel('Densidade')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Gráfico de Regressão: 'Preço' vs 'Nota'
# Escolha das variáveis: 'Preço' e 'Nota' são variáveis numéricas onde queremos entender a relação linear entre elas. O regplot adiciona uma linha de regressão linear para visualizar essa tendência e a dispersão dos pontos em torno dela.
# Biblioteca e Método: O gráfico de regressão é gerado usando a biblioteca `seaborn` com o método `regplot`.
column_reg_x = 'Preço'
column_reg_y = 'Nota' # Corrigido de 'Avaliação' para 'Nota'

plt.figure(figsize=(10, 6))
sns.regplot(x=df_new[column_reg_x], y=df_new[column_reg_y], scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title(f'Gráfico de Regressão: {column_reg_x} vs. {column_reg_y}')
plt.xlabel(column_reg_x)
plt.ylabel(column_reg_y)
plt.grid(True, alpha=0.75)
plt.show()
```
