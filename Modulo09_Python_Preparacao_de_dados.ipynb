## Engenharia de Features

A **Engenharia de Features** é o processo de usar o conhecimento do domínio para criar variáveis (features) a partir de dados brutos que tornam os algoritmos de aprendizado de máquina mais eficazes. Basicamente, é transformar os dados existentes para extrair informações mais relevantes para o modelo.

### Principais Ferramentas e Técnicas:

1.  **Transformação de Variáveis Categóricas:**
    *   **One-Hot Encoding:** Converte variáveis categóricas em um formato numérico que pode ser fornecido a algoritmos de ML. Cria novas colunas indicando a presença (1) ou ausência (0) de cada categoria.
    *   **Label Encoding:** Atribui um valor numérico único a cada categoria. Pode ser útil para variáveis ordinais, mas cuidado para não introduzir uma ordem artificial.

2.  **Tratamento de Dados Numéricos:**
    *   **Padronização (Standardization - StandardScaler):** Ajusta os dados para que tenham média zero e desvio padrão um. Útil para algoritmos que são sensíveis à escala das features (ex: SVM, Regressão Logística).
    *   **Normalização (Normalization - MinMaxScaler):** Redimensiona os dados para um intervalo fixo, geralmente entre 0 e 1. Útil para algoritmos que usam distâncias como medida (ex: KNN).
    *   **Binning (Discretização):** Converte variáveis numéricas contínuas em variáveis categóricas, dividindo-as em "caixas" ou intervalos.

3.  **Criação de Novas Features:**
    *   **Interações de Features:** Combinar duas ou mais features existentes para criar uma nova. Por exemplo, "idade x salário" ou "comprimento x largura".
    *   **Features Polinomiais:** Criar potências de features existentes (ex: x, x², x³). Ajuda a capturar relações não lineares.
    *   **Features Baseadas em Tempo:** Extrair informações de datas e horas, como dia da semana, mês, ano, trimestre, feriado, etc.
    *   **Agregações:** Resumir dados por grupos (ex: média de vendas por produto, contagem de eventos por usuário).

4.  **Redução de Dimensionalidade:**
    *   **Análise de Componentes Principais (PCA):** Transforma um conjunto de features correlacionadas em um conjunto menor de componentes não correlacionados (componentes principais), mantendo a maior parte da variância dos dados.
    *   **Seleção de Features (Feature Selection):** Escolher um subconjunto das features mais relevantes para o modelo, descartando as menos importantes. Pode ser feito por métodos estatísticos (ANOVA, Qui-quadrado), modelos (Embedded Methods) ou wrappers (RFE).

5.  **Tratamento de Valores Ausentes:**
    *   **Imputação:** Preencher valores ausentes com a média, mediana, moda, ou usando modelos preditivos.
    *   **Remoção:** Remover linhas ou colunas que contêm valores ausentes (com cautela, para não perder dados importantes).

As ferramentas mais comuns para realizar a engenharia de features em Python incluem:

*   **Pandas:** Para manipulação e transformação de dados em DataFrames.
*   **Scikit-learn (sklearn):** Oferece uma vasta gama de pré-processadores, transformadores e seletores de features (ex: `StandardScaler`, `MinMaxScaler`, `OneHotEncoder`, `LabelEncoder`, `PolynomialFeatures`, `PCA`, `SelectKBest`).
*   **NumPy:** Para operações numéricas eficientes.
*   **Feature-engine:** Uma biblioteca focada exclusivamente em engenharia de features, oferecendo transformadores mais avançados e personalizáveis.

A engenharia de features é muitas vezes a etapa mais importante no desenvolvimento de um modelo de ML, pois a qualidade e relevância das features impactam diretamente o desempenho do modelo.

### 1. Label Encoding

**O que é?**

`Label Encoding` é uma técnica de pré-processamento usada para converter rótulos categóricos em um formato numérico. Cada categoria única em uma coluna é atribuída a um número inteiro exclusivo.

**Quando usar?**

*   É mais apropriado para variáveis categóricas ordinais, ou seja, aquelas que possuem uma ordem intrínseca (ex: 'Pequeno', 'Médio', 'Grande').
*   Pode ser usado para variáveis nominais (sem ordem) quando o número de categorias é muito grande e `One-Hot Encoding` criaria muitas colunas, mas com a ressalva de que pode introduzir uma ordem artificial que o modelo pode interpretar erroneamente.

**Vantagens:**

*   Simples de implementar.
*   Não aumenta a dimensionalidade do **dataset** (conjunto de dados - uma coleção de dados relacionados, geralmente organizada em tabelas) (ao contrário do `One-Hot Encoding`).

**Desvantagens:**

*   Para variáveis nominais, a atribuição de números inteiros pode levar o modelo a assumir uma relação ordinal ou de magnitude entre as categorias, o que pode prejudicar o desempenho do modelo se essa ordem não existir na realidade.

## Resumo de Engenharia de Features e Análise Exploratória de Dados (EDA)

### Engenharia de Features

A **Engenharia de Features** é o processo de transformar dados brutos em variáveis mais úteis para modelos de Machine Learning, melhorando seu desempenho e interpretabilidade. Isso envolve a **transformação** e **categorização** dos dados.

#### 1. Categorização e Codificação de Dados Categóricos

*   **Label Encoding:** Atribui um número inteiro único a cada categoria. Ideal para variáveis **ordinais** (com ordem intrínseca). Pode introduzir ordem artificial para dados nominais.
*   **OrdinalEncoder:** Semelhante ao Label Encoding, mas permite definir explicitamente a ordem das categorias para variáveis **ordinais**, preservando a hierarquia.
*   **One-Hot Encoding:** Cria colunas binárias (0 ou 1) para cada categoria. Essencial para variáveis **nominais** (sem ordem) para evitar ordem artificial, mas aumenta a dimensionalidade do dataset.

#### 2. Transformação e Escalonamento de Dados Numéricos

*   **Normalização (MinMaxScaler):** Redimensiona valores numéricos para um intervalo fixo (geralmente 0 a 1). É uma técnica de **normalização**, útil para algoritmos sensíveis à escala, mas vulnerável a outliers.
*   **Padronização (StandardScaler):** Transforma dados para ter média 0 e desvio padrão 1. É uma técnica de **padronização**, ideal para algoritmos que assumem distribuição gaussiana ou são sensíveis à escala, sendo menos sensível a outliers que o MinMaxScaler.
*   **RobustScaler:** Escala usando mediana e Intervalo Interquartil (IQR). É uma técnica de **escalonamento** mais robusta a outliers, ideal para distribuições assimétricas ou com valores extremos.

#### 3. Transformação de Distribuições

*   **Transformação Logarítmica:** Aplica a função logaritmo para **reduzir assimetria** em distribuições inclinadas (ex: renda) e estabilizar a variância. Ajuda a aproximar os dados de uma distribuição normal. Não aplicável diretamente a zeros ou valores negativos.
*   **Transformação Box-Cox:** Uma transformação mais flexível que a logarítmica, que pode **transformar dados em uma distribuição mais próxima da normal**. Encontra o valor ideal de λ (lambda) que minimiza a assimetria e estabiliza a variância. **Aplica-se apenas a dados estritamente positivos.** É uma transformação de potência que generaliza a transformação logarítmica (quando λ=0).

#### 4. Tratamento de Valores Ausentes

*   **Remoção:** Excluir linhas ou colunas com valores faltantes. Método simples, mas pode causar perda de dados significativa.
*   **Imputação (Média, Mediana, Moda):** Preencher valores ausentes com estimativas: média para dados numéricos (sensível a outliers), mediana para dados numéricos (mais robusta a outliers) e moda para dados categóricos ou numéricos.

### Análise Exploratória de Dados (EDA)

A **Análise Exploratória de Dados (EDA)** é a fase de investigação inicial para entender as características principais de um conjunto de dados. Ela usa principalmente métodos visuais e estatísticos para descobrir padrões, identificar anomalias, testar hipóteses e validar suposições.

#### Principais Objetivos da EDA:

*   **Compreender as Distribuições:** Analisar a forma, centralidade e dispersão das variáveis usando gráficos como histogramas e box plots.
*   **Detectar Outliers:** Identificar valores atípicos que podem distorcer análises e modelos.
*   **Explorar Relações:** Entender a correlação e dependência entre diferentes variáveis através de gráficos de dispersão e matrizes de correlação.
*   **Identificar Padrões e Tendências:** Descobrir comportamentos, sazonalidade ou ciclos nos dados.
*   **Guiar a Pré-processamento:** Fornecer insights para a engenharia de features, tratamento de valores ausentes e limpeza de dados.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Exemplo de DataFrame com uma coluna categórica
data = {'Tamanho': ['Pequeno', 'Médio', 'Grande', 'Pequeno', 'Grande', 'Médio']}
df_label = pd.DataFrame(data)

print("DataFrame Original:")
display(df_label)

# Inicializa o LabelEncoder
le = LabelEncoder()

# Aplica o Label Encoding à coluna 'Tamanho'
df_label['Tamanho_Encoded'] = le.fit_transform(df_label['Tamanho'])

print("\nDataFrame após Label Encoding:")
display(df_label)

print("\nClasses mapeadas:")
for i, item in enumerate(le.classes_):
    print(f"{item} -> {i}")
```

### 2. Normalização (MinMaxScaler)

**O que é?**

`MinMaxScaler` é uma técnica de normalização que redimensiona os dados para um intervalo fixo, geralmente entre 0 e 1. Ele faz isso subtraindo o valor mínimo de cada feature e dividindo pelo intervalo (máximo - mínimo) da feature. A fórmula é:

`X_normalized = (X - X_min) / (X_max - X_min)`

**Quando usar?**

*   É muito útil para algoritmos que são sensíveis à escala dos dados, como redes neurais, SVMs (Support Vector Machines) com kernels baseados em distância e algoritmos baseados em distância como KNN (K-Nearest Neighbors).
*   Quando você deseja que todas as features contribuam igualmente para a distância calculada, independentemente de suas unidades ou magnitudes originais.
*   Quando a distribuição dos dados não é gaussiana ou quando a presença de **outliers** (valores atípicos - pontos de dados que se desviam significativamente de outras observações) é um fator, mas você deseja manter o formato da distribuição original (apenas redimensioná-la).

**Vantagens:**

*   Garantia de que todas as features estejam na mesma escala, o que pode melhorar o desempenho de muitos algoritmos de ML.
*   É simples de entender e implementar.
*   Preserva a forma da distribuição original dos dados.

**Desvantagens:**

*   É muito sensível a **outliers** (valores atípicos). Um único outlier pode distorcer significativamente o intervalo (Min-Max) e comprimir os dados "normais" em um intervalo muito pequeno.
*   Se houver novos dados fora do intervalo original (Min-Max), o scaler precisará ser reajustado, ou os novos dados não serão escalados corretamente.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Exemplo de DataFrame com duas colunas numéricas em escalas diferentes
data = {
    'Idade': [25, 30, 35, 40, 60, 22, 55],
    'Salario_Anual': [30000, 45000, 50000, 70000, 120000, 28000, 90000]
}
df_minmax = pd.DataFrame(data)

print("DataFrame Original:")
display(df_minmax)

# Inicializa o MinMaxScaler
scaler = MinMaxScaler()

# Aplica a normalização às colunas 'Idade' e 'Salario_Anual'
df_minmax_scaled = df_minmax.copy()
df_minmax_scaled[['Idade_Scaled', 'Salario_Anual_Scaled']] = scaler.fit_transform(df_minmax[['Idade', 'Salario_Anual']])

print("\nDataFrame após MinMaxScaler:")
display(df_minmax_scaled)

print("\nValores mínimos e máximos após escalonamento:")
print(f"Idade Scaled: Min={df_minmax_scaled['Idade_Scaled'].min():.2f}, Max={df_minmax_scaled['Idade_Scaled'].max():.2f}")
print(f"Salario Anual Scaled: Min={df_minmax_scaled['Salario_Anual_Scaled'].min():.2f}, Max={df_minmax_scaled['Salario_Anual_Scaled'].max():.2f}")
```

### 3. One-Hot Encoding

**O que é?**

`One-Hot Encoding` é uma técnica de pré-processamento usada para converter variáveis categóricas nominais (aquelas sem ordem intrínseca) em um formato numérico. Para cada categoria única em uma coluna, ele cria uma nova coluna binária (dummy variable) que contém 1 se a linha pertence a essa categoria e 0 caso contrário.

**Quando usar?**

*   É a técnica mais comum para variáveis categóricas nominais (ex: 'Vermelho', 'Azul', 'Verde'; 'Carro', 'Moto', 'Bicicleta').
*   É preferível ao `Label Encoding` para variáveis nominais, pois evita que o modelo interprete uma relação ordinal ou de magnitude entre as categorias, o que não existe na realidade.
*   Útil para algoritmos que esperam entrada numérica e assumem que valores maiores indicam uma ordem (ex: regressão linear, árvores de decisão).

**Vantagens:**

*   Não introduz uma ordem artificial entre as categorias.
*   Preserva a individualidade de cada categoria.
*   Modelos de ML podem interpretar facilmente as novas features binárias.

**Desvantagens:**

*   Aumenta a dimensionalidade do **dataset** (conjunto de dados), especialmente se a variável categórica tiver muitas categorias únicas. Isso pode levar ao problema de "maldição da dimensionalidade" e tornar o modelo mais lento ou propenso a **overfitting** (superajuste - quando um modelo de aprendizado de máquina aprende os detalhes e o ruído dos dados de treinamento a ponto de ter um desempenho ruim em dados novos e não vistos).
*   Pode causar multicolinearidade (quando uma coluna pode ser perfeitamente prevista por outras colunas), o que pode ser um problema para alguns modelos de regressão. Para mitigar isso, uma das colunas dummy pode ser removida (drop_first=True).

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Exemplo de DataFrame com uma coluna categórica nominal
data = {'Cor': ['Vermelho', 'Azul', 'Verde', 'Vermelho', 'Azul', 'Amarelo']}
df_onehot = pd.DataFrame(data)

print("DataFrame Original:")
display(df_onehot)

# Inicializa o OneHotEncoder
# handle_unknown='ignore' para lidar com categorias desconhecidas em dados futuros sem erro
# sparse_output=False para retornar um array denso do numpy (facilita a conversão para DataFrame)
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Aplica o One-Hot Encoding à coluna 'Cor'
# Cria um DataFrame a partir do resultado do encoding
df_encoded_cols = pd.DataFrame(encoder.fit_transform(df_onehot[['Cor']]),
                               columns=encoder.get_feature_names_out(['Cor']))

# Concatena as colunas codificadas com o DataFrame original
df_onehot_encoded = pd.concat([df_onehot, df_encoded_cols], axis=1)

print("\nDataFrame após One-Hot Encoding:")
display(df_onehot_encoded)

print("\nCategorias codificadas:")
print(encoder.get_feature_names_out(['Cor']))
```

### Análise Prática com One-Hot Encoding

Depois de aplicar o One-Hot Encoding, como visto no exemplo acima, a coluna categórica original ('Cor' em nosso caso) é substituída por múltiplas colunas binárias (`Cor_Amarelo`, `Cor_Azul`, etc.).

**O que essas novas colunas significam para a análise?**

Cada uma dessas novas colunas representa a presença (valor 1) ou ausência (valor 0) de uma categoria específica. Por exemplo:

*   Se uma linha tem `Cor_Amarelo = 1`, significa que a observação original pertencia à categoria 'Amarelo'.
*   Se `Cor_Azul = 0`, significa que a observação não era da categoria 'Azul'.

**Como isso é usado na análise (e que resultados busca)?**

1.  **Entrada para Modelos de Machine Learning:** A maioria dos algoritmos de ML, especialmente modelos lineares (Regressão Linear, Regressão Logística, SVMs, Redes Neurais), exigem entradas numéricas. O One-Hot Encoding transforma as categorias em um formato que esses modelos podem processar.

2.  **Evitar a Falsa Ordinalidade:** Para variáveis nominais (sem ordem intrínseca, como 'Cor'), o One-Hot Encoding impede que o modelo atribua uma ordem ou peso artificial às categorias. Se tivéssemos usado Label Encoding para 'Cor' (ex: Amarelo=0, Azul=1, Verde=2, Vermelho=3), o modelo poderia interpretar erroneamente que 'Vermelho' é 'maior' ou 'mais importante' que 'Amarelo', o que não é verdade para cores.

3.  **Permitir que o Modelo Aprenda Pesos Individuais:** Com colunas binárias separadas, o modelo pode aprender um "peso" ou "coeficiente" independente para cada categoria. Por exemplo, em um modelo de previsão de preços de carros, se a cor 'Vermelho' for geralmente associada a um preço mais alto, o modelo poderá aprender um coeficiente positivo para a coluna `Cor_Vermelho`, sem que isso afete o peso de `Cor_Azul`.

4.  **Interpretabilidade (em alguns casos):** Em modelos como regressão linear, os coeficientes das colunas codificadas podem indicar a influência de cada categoria na variável alvo, facilitando a interpretação. Por exemplo, um coeficiente positivo alto para `Cor_Vermelho` poderia sugerir que carros vermelhos tendem a ter um valor maior.

**Exemplo Hipotético:**

Imagine que estamos construindo um modelo para prever a satisfação do cliente com base em diversas features, incluindo a 'Cor' do produto. Após o One-Hot Encoding, as colunas `Cor_Amarelo`, `Cor_Azul`, `Cor_Verde`, `Cor_Vermelho` seriam passadas para o modelo. O objetivo seria:

*   **Resultados Desejados:** O modelo conseguiria identificar se alguma cor específica (por exemplo, 'Azul') está estatisticamente mais associada a uma maior ou menor satisfação do cliente, sem que essa relação seja confundida com a de outras cores devido a uma codificação numérica arbitrária.

Em resumo, o One-Hot Encoding é uma técnica crucial para garantir que os modelos de ML interpretem corretamente as informações categóricas, melhorando sua capacidade preditiva e a validade de suas inferências.

### 4. RobustScaler

**O que é?**

`RobustScaler` é uma técnica de escalonamento que dimensiona os dados usando o intervalo interquartil (IQR), que é a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1). Ele remove a mediana e escala os dados com base no IQR, tornando-o mais robusto a **outliers** (valores atípicos - pontos de dados que se desviam significativamente de outras observações) do que o `MinMaxScaler` ou `StandardScaler`.

A fórmula é:
`X_scaled = (X - Mediana) / IQR`

**Quando usar?**

*   É ideal para **datasets** (conjuntos de dados) que contêm muitos **outliers** (valores atípicos), pois a mediana e o IQR são menos afetados por valores extremos do que a média e o desvio padrão.
*   Quando a distribuição dos dados não é normal ou possui **caudas pesadas** (uma distribuição de dados que tem mais valores nas "caudas" - extremos - do que uma distribuição normal, indicando a presença de muitos valores extremos).
*   Assim como outros escalonadores, é útil para algoritmos que são sensíveis à escala das features.

**Vantagens:**

*   É robusto a **outliers** (valores atípicos), ou seja, menos impactado por valores extremos no **dataset** (conjunto de dados).
*   Preserva a forma da distribuição original dos dados, mas a torna mais compacta em torno da mediana.

**Desvantagens:**

*   Pode não produzir resultados tão bons quanto o `StandardScaler` ou `MinMaxScaler` se o **dataset** (conjunto de dados) não tiver **outliers** (valores atípicos) ou se a distribuição for aproximadamente normal, pois o IQR pode ser menos informativo do que o desvio padrão nesse cenário.
*   Os valores escalonados não ficam restritos a um intervalo específico (como [0, 1] no `MinMaxScaler`) ou a terem média zero e desvio padrão um exatos (como no `StandardScaler`), o que pode ser uma desvantagem para alguns algoritmos ou requisitos específicos.

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler
import numpy as np

# Exemplo de DataFrame com outliers
data = {
    'Tempo_Atividade': [10, 15, 20, 25, 30, 100, 12, 18],
    'Valor_Transacao': [100, 150, 200, 250, 300, 5000, 120, 180]
}
df_robust = pd.DataFrame(data)

print("DataFrame Original:")
display(df_robust)

# Inicializa o RobustScaler
scaler = RobustScaler()

# Aplica o escalonamento robusto às colunas
df_robust_scaled = df_robust.copy()
df_robust_scaled[['Tempo_Atividade_Scaled', 'Valor_Transacao_Scaled']] = scaler.fit_transform(df_robust[['Tempo_Atividade', 'Valor_Transacao']])

print("\nDataFrame após RobustScaler:")
display(df_robust_scaled)

print("\nEstatísticas após escalonamento:")
print(f"Tempo_Atividade_Scaled: Mediana={df_robust_scaled['Tempo_Atividade_Scaled'].median():.2f}, IQR={df_robust_scaled['Tempo_Atividade_Scaled'].quantile(0.75) - df_robust_scaled['Tempo_Atividade_Scaled'].quantile(0.25):.2f}")
print(f"Valor_Transacao_Scaled: Mediana={df_robust_scaled['Valor_Transacao_Scaled'].median():.2f}, IQR={df_robust_scaled['Valor_Transacao_Scaled'].quantile(0.75) - df_robust_scaled['Valor_Transacao_Scaled'].quantile(0.25):.2f}")
```

### 5. Padronização (StandardScaler)

**O que é?**

`StandardScaler` é uma técnica de **padronização** (Standardization) que transforma os **dados** (data) para que tenham uma **média** (mean) de 0 e um **desvio padrão** (standard deviation) de 1. Ele faz isso subtraindo a média da **feature** (característica) e dividindo pelo desvio padrão da feature. A fórmula é:

`X_scaled = (X - média) / desvio_padrão`

**Quando usar?**

*   É muito útil para **algoritmos de aprendizado de máquina** (machine learning algorithms) que assumem que os **dados** (data) seguem uma **distribuição Gaussiana** (Gaussian distribution) ou que são sensíveis à escala das **features** (características), como **modelos lineares** (linear models - e.g., Regressão Linear, Regressão Logística), **redes neurais** (neural networks) e **Máquinas de Vetores de Suporte (SVMs)** (Support Vector Machines).
*   Quando as **features** (características) têm unidades de medida diferentes ou amplitudes muito variadas.

**Vantagens:**

*   Garante que todas as **features** (características) contribuam igualmente para o modelo, evitando que **features** (características) com valores maiores dominem o processo de aprendizado.
*   É menos sensível a **outliers** (valores atípicos - pontos de dados que se desviam significativamente de outras observações) do que o `MinMaxScaler`, pois usa a média e o desvio padrão, que são afetados, mas não tão drasticamente quanto o min-max.

**Desvantagens:**

*   Altera a distribuição dos **dados** (data) e não mantém os valores dentro de um intervalo específico (como o `MinMaxScaler` faz entre 0 e 1).
*   Para **datasets** (conjuntos de dados) com muitos **outliers** (valores atípicos), a média e o desvio padrão podem ser distorcidos, levando a um escalonamento menos eficaz. Nesses casos, o `RobustScaler` pode ser uma alternativa melhor.

**Exemplo:**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np

# Exemplo de DataFrame com duas colunas numéricas em escalas e distribuições diferentes
data = {
    'Altura_cm': [160, 175, 182, 165, 190, 170, 178],
    'Peso_kg': [55, 70, 85, 60, 95, 68, 77]
}
df_standard = pd.DataFrame(data)

print("DataFrame Original:")
display(df_standard)

# Inicializa o StandardScaler
scaler = StandardScaler()

# Aplica a padronização às colunas 'Altura_cm' e 'Peso_kg'
df_standard_scaled = df_standard.copy()
df_standard_scaled[['Altura_Scaled', 'Peso_Scaled']] = scaler.fit_transform(df_standard[['Altura_cm', 'Peso_kg']])

print("\nDataFrame após StandardScaler:")
display(df_standard_scaled)

print("\nEstatísticas após escalonamento:")
print(f"Altura Scaled: Média={df_standard_scaled['Altura_Scaled'].mean():.2f}, Desvio Padrão={df_standard_scaled['Altura_Scaled'].std():.2f}")
print(f"Peso Scaled: Média={df_standard_scaled['Peso_Scaled'].mean():.2f}, Desvio Padrão={df_standard_scaled['Peso_Scaled'].std():.2f}")
```

### 6. Codificação Ordinal (OrdinalEncoder)

**O que é?**

`OrdinalEncoder` é uma técnica usada para transformar variáveis categóricas ordinais em representações numéricas inteiras. Diferente do `LabelEncoder`, o `OrdinalEncoder` permite especificar a ordem das categorias explicitamente, o que é crucial para variáveis onde a relação de ordem é importante e precisa ser mantida. Cada categoria é mapeada para um número inteiro com base na ordem definida.

**Quando usar?**

*   É ideal para variáveis categóricas que possuem uma ordem intrínseca clara, onde o "maior" ou "menor" tem significado (ex: 'Baixo', 'Médio', 'Alto'; 'Péssimo', 'Ruim', 'Bom', 'Ótimo').
*   Quando o modelo de aprendizado de máquina precisa entender essa relação de ordem entre as categorias.

**Vantagens:**

*   Preserva a relação de ordem entre as categorias, o que pode ser benéfico para alguns algoritmos (como árvores de decisão e modelos lineares, se a relação for linear).
*   Não aumenta a dimensionalidade do dataset.
*   Permite controle explícito sobre o mapeamento da ordem das categorias.

**Desvantagens:**

*   A definição da ordem das categorias é manual e pode ser um desafio se não houver uma ordem clara ou se for ambígua.
*   Se usada para variáveis nominais (sem ordem), pode introduzir uma ordem artificial que pode confundir o modelo, da mesma forma que o `LabelEncoder`.

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Exemplo de DataFrame com uma coluna categórica ordinal (nível de escolaridade)
data = {
    'Nivel_Escolaridade': ['Ensino Médio', 'Graduação', 'Pós-Graduação', 'Ensino Médio', 'Doutorado', 'Graduação'],
    'Idade': [20, 25, 30, 22, 35, 28]
}
df_ordinal = pd.DataFrame(data)

print("DataFrame Original:")
display(df_ordinal)

# Definir a ordem esperada das categorias
# É crucial que esta lista contenha todas as categorias únicas na ordem correta
categories_order = [['Ensino Médio', 'Graduação', 'Pós-Graduação', 'Doutorado']]

# Inicializa o OrdinalEncoder com a ordem definida
encoder = OrdinalEncoder(categories=categories_order)

# Aplica a codificação ordinal à coluna 'Nivel_Escolaridade'
df_ordinal['Nivel_Escolaridade_Encoded'] = encoder.fit_transform(df_ordinal[['Nivel_Escolaridade']])

print("\nDataFrame após Ordinal Encoding:")
display(df_ordinal)

print("\nClasses mapeadas e sua ordem:")
for i, category in enumerate(categories_order[0]):
    print(f"{category} -> {i}")
```

### 7. Transformação Logarítmica

**O que é?**
A **Transformação Logarítmica** é uma técnica matemática que aplica a função logaritmo (geralmente log base `e` - logaritmo natural, ou log base 10) aos valores de uma feature. As transformações logarítmicas são frequentemente usadas para modelar dados que exibem crescimento exponencial ou que estão fortemente inclinados (skewed).

A forma mais comum é `log(x)` ou, para lidar com valores zero, `log(1 + x)` (conhecido como `log1p` em bibliotecas como o NumPy).

**Quando usar?**
*   **Reduzir Assimetria (Skewness):** É extremamente eficaz para transformar distribuições de dados altamente assimétricas (especialmente assimétricas à direita, onde há uma cauda longa de valores altos). Variáveis como renda, preço de imóveis, ou volume de vendas muitas vezes seguem essa distribuição.
*   **Estabilizar Variância:** Pode ajudar a estabilizar a variância de dados, o que é útil em modelos que assumem homocedasticidade (variância constante).
*   **Lidar com Relações Multiplicativas:** Transforma relações multiplicativas em aditivas, o que pode simplificar a modelagem. Por exemplo, se `y = a * x^b`, `log(y) = log(a) + b * log(x)`, tornando a relação linear.
*   **Reduzir o Impacto de Outliers:** Valores extremos são 'comprimidos' pela função logarítmica, diminuindo seu impacto nos modelos.

**Vantagens:**
*   Normaliza a distribuição dos dados, aproximando-a de uma distribuição normal (gaussiana), o que é um pressuposto para muitos modelos estatísticos e de ML.
*   Torna os dados mais simétricos e menos suscetíveis a outliers.
*   Pode melhorar o desempenho de modelos de regressão linear que assumem linearidade e normalidade dos resíduos.

**Desvantagens:**
*   Não pode ser aplicada diretamente a valores zero ou negativos (a não ser que se use `log(1+x)` para zeros ou outras transformações para negativos).
*   A interpretação dos coeficientes do modelo em uma escala logarítmica é menos intuitiva do que na escala original.
*   Pode não ser a melhor escolha se a distribuição dos dados não for assimétrica ou se a relação com a variável alvo for linear na escala original.

### Caso Prático: Transformando Dados de Renda Assimétricos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Criar um DataFrame com dados de renda altamente assimétricos (simulando a realidade)
data = {
    'Renda_Mensal': np.random.exponential(scale=1500, size=1000) + 500 # Adiciona 500 para evitar zeros e simular renda mínima
}
df_log = pd.DataFrame(data)

print("DataFrame Original (primeiras 5 linhas):")
display(df_log.head())

# Plotar a distribuição original da Renda_Mensal
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_log['Renda_Mensal'], kde=True)
plt.title('Distribuição Original da Renda Mensal')
plt.xlabel('Renda Mensal')
plt.ylabel('Frequência')

# Aplicar a transformação logarítmica (usando log1p para lidar com possíveis valores próximos de zero)
df_log['Renda_Mensal_Log'] = np.log1p(df_log['Renda_Mensal'])

# Plotar a distribuição da Renda_Mensal após a transformação logarítmica
plt.subplot(1, 2, 2)
sns.histplot(df_log['Renda_Mensal_Log'], kde=True)
plt.title('Distribuição da Renda Mensal (Transformação Logarítmica)')
plt.xlabel('Log(1 + Renda Mensal)')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()

print("\nEstatísticas da Renda Mensal antes da transformação:")
print(df_log['Renda_Mensal'].describe())
print(f"Assimetria (Skewness) original: {df_log['Renda_Mensal'].skew():.2f}")

print("\nEstatísticas da Renda Mensal após a transformação logarítmica:")
print(df_log['Renda_Mensal_Log'].describe())
print(f"Assimetria (Skewness) após log: {df_log['Renda_Mensal_Log'].skew():.2f}")

print("\nDataFrame após Transformação Logarítmica (primeiras 5 linhas):")
display(df_log.head())
```

### 8. Tratamento de Valores Ausentes (Missing Values)

Valores ausentes são um problema comum em **datasets** (conjuntos de dados) reais e podem afetar negativamente a performance de modelos de Machine Learning. É crucial identificar e tratá-los adequadamente.

#### Principais Técnicas:

1.  **Remoção (Deletion):**
    *   **Remoção de Linhas (Row-wise Deletion / Listwise Deletion):** Remove-se a linha inteira que contém valores ausentes. É simples, mas pode resultar na perda significativa de dados se muitas linhas tiverem valores ausentes, o que pode levar a um **viés** (bias) nos dados restantes.
    *   **Remoção de Colunas (Column-wise Deletion):** Remove-se uma coluna inteira se ela tiver uma grande proporção de valores ausentes (ex: mais de 50-70%). Isso é feito quando a feature não é considerada informativa devido à quantidade de dados faltantes.
    
    **Quando usar:**
    *   Quando o número de valores ausentes é pequeno em relação ao tamanho total do dataset.
    *   Quando a perda de algumas linhas ou colunas não compromete a representatividade ou o tamanho do dataset.

2.  **Imputação (Imputation):**
    *   **Imputação por Constante:** Substituir valores ausentes por uma constante (ex: 0, -1, 'Desconhecido'). Útil para variáveis categóricas ou quando a ausência de um valor é significativa por si só.
    *   **Imputação por Média/Mediana/Moda:**
        *   **Média:** Valores ausentes em colunas numéricas são preenchidos com a média dos valores existentes na coluna. Simples, mas sensível a **outliers** (valores atípicos).
        *   **Mediana:** Valores ausentes em colunas numéricas são preenchidos com a mediana. Mais robusta a outliers que a média.
        *   **Moda:** Valores ausentes em colunas numéricas ou categóricas são preenchidos com o valor mais frequente (moda) na coluna.
    *   **Imputação por Regressão/Modelo Preditivo:** Usar outros recursos do dataset para prever os valores ausentes de uma variável. Mais sofisticado, mas pode ser complexo e demorado.
    *   **Imputação Múltipla:** Gerar múltiplos datasets preenchidos com diferentes imputações e combinar os resultados das análises. Ajuda a quantificar a incerteza da imputação.
    
    **Quando usar:**
    *   Quando remover os dados resultaria em uma perda substancial de informações.
    *   Quando os valores ausentes são aleatórios e não indicam um padrão específico.

```python
import pandas as pd
import numpy as np

# Criar um DataFrame com valores ausentes
data = {
    'Idade': [25, 30, np.nan, 40, 60, 22, np.nan],
    'Salario': [3000, np.nan, 5000, 7000, 12000, 2800, 9000],
    'Cidade': ['SP', 'RJ', 'BH', 'SP', np.nan, 'RJ', 'BH']
}
df_missing = pd.DataFrame(data)

print("DataFrame Original com Valores Ausentes:")
display(df_missing)

# Verificar a quantidade de valores ausentes por coluna
print("\nValores ausentes por coluna:")
print(df_missing.isnull().sum())

# --- Exemplo de Imputação ---

# 1. Imputação por Mediana para 'Idade' (numérica, mais robusta a outliers)
df_imputed_median = df_missing.copy()
median_age = df_imputed_median['Idade'].median()
df_imputed_median['Idade'] = df_imputed_median['Idade'].fillna(median_age)

# 2. Imputação por Média para 'Salario' (numérica, se não houver outliers significativos)
df_imputed_mean = df_imputed_median.copy() # Continuar do DataFrame com Idade imputada
mean_salary = df_imputed_mean['Salario'].mean()
df_imputed_mean['Salario'] = df_imputed_mean['Salario'].fillna(mean_salary)

# 3. Imputação por Moda para 'Cidade' (categórica)
df_imputed_mode = df_imputed_mean.copy() # Continuar do DataFrame com Idade e Salario imputados
mode_city = df_imputed_mode['Cidade'].mode()[0] # .mode() retorna uma Series, pegamos o primeiro valor
df_imputed_mode['Cidade'] = df_imputed_mode['Cidade'].fillna(mode_city)

print("\nDataFrame após Imputação (Mediana para Idade, Média para Salário, Moda para Cidade):")
display(df_imputed_mode)

print("\nValores ausentes após Imputação:")
print(df_imputed_mode.isnull().sum())

print("\nEstatísticas após Imputação:")
print(f"Idade - Média: {df_imputed_mode['Idade'].mean():.2f}, Mediana: {df_imputed_mode['Idade'].median():.2f}, Moda: {df_imputed_mode['Idade'].mode()[0]:.2f}")
print(f"Salario - Média: {df_imputed_mode['Salario'].mean():.2f}, Mediana: {df_imputed_mode['Salario'].median():.2f}, Moda: {df_imputed_mode['Salario'].mode()[0]:.2f}")
print(f"Cidade - Moda: {df_imputed_mode['Cidade'].mode()[0]}")

# --- Exemplo de Remoção (com cautela!) ---

# Criar um novo DataFrame para demonstração de remoção para não alterar o anterior
df_removed = pd.DataFrame(data)

print("\nDataFrame Original para Remoção:")
display(df_removed)

# Remover linhas que contêm QUALQUER valor ausente
# Use com muita cautela, pois pode remover muitos dados!
df_removed_rows = df_removed.dropna()

print("\nDataFrame após Remoção de Linhas com Valores Ausentes:")
display(df_removed_rows)

print(f"Linhas removidas: {len(df_removed) - len(df_removed_rows)}")
```

## Resumo de Engenharia de Features e Análise Exploratória de Dados (EDA)

### Engenharia de Features

A **Engenharia de Features** é o processo de preparar e transformar dados brutos em variáveis mais eficazes para modelos de Machine Learning, melhorando seu desempenho e interpretabilidade.

#### 1. Categorização e Codificação de Dados Categóricos

*   **Label Encoding:** Transforma categorias ordinais (com ordem, ex: Pequeno, Médio) em números inteiros únicos. Pode introduzir ordem artificial em dados nominais.
*   **OrdinalEncoder:** Similar ao Label Encoding, mas permite definir manualmente a ordem das categorias para variáveis ordinais, garantindo a preservação da hierarquia.
*   **One-Hot Encoding:** Converte categorias nominais (sem ordem, ex: Vermelho, Azul) em colunas binárias (0 ou 1). Evita a criação de uma ordem artificial, mas aumenta a dimensionalidade do dataset.

#### 2. Transformação e Escalonamento de Dados Numéricos

*   **Normalização (MinMaxScaler):** Redimensiona valores numéricos para um intervalo fixo (geralmente 0 a 1). Útil para algoritmos sensíveis à escala, mas vulnerável a outliers.
*   **Padronização (StandardScaler):** Transforma dados para ter média 0 e desvio padrão 1. Adequado para modelos que assumem distribuição Gaussiana e menos sensível a outliers que o MinMaxScaler.
*   **RobustScaler:** Escala dados usando a mediana e o Intervalo Interquartil (IQR). É o mais robusto a outliers, ideal para distribuições assimétricas ou com valores extremos.

#### 3. Transformação de Distribuições

*   **Transformação Logarítmica:** Aplica a função logaritmo para reduzir a assimetria (skewness) em distribuições inclinadas (ex: renda) e estabilizar a variância. Ajuda a aproximar os dados de uma distribuição normal. Não aplicável diretamente a zeros ou valores negativos.

#### 4. Tratamento de Dados Ausentes

*   **Remoção:** Excluir linhas ou colunas com dados faltantes. Método simples, mas pode resultar em perda significativa de informações.
*   **Imputação (Média, Mediana, Moda):** Preencher valores ausentes com estimativas: média para dados numéricos (sensível a outliers), mediana para dados numéricos (mais robusta a outliers) e moda para dados categóricos ou numéricos.

### Análise Exploratória de Dados (EDA)

A **Análise Exploratória de Dados (EDA)** é a fase de investigação inicial para entender as características principais de um conjunto de dados. Ela usa principalmente métodos visuais e estatísticos para descobrir padrões, identificar anomalias, testar hipóteses e validar suposições.

#### Principais Objetivos da EDA:

*   **Compreender as Distribuições:** Analisar a forma, centralidade e dispersão das variáveis usando gráficos como histogramas e box plots.
*   **Detectar Outliers:** Identificar valores atípicos que podem distorcer análises e modelos.
*   **Explorar Relações:** Entender a correlação e dependência entre diferentes variáveis através de gráficos de dispersão e matrizes de correlação.
*   **Identificar Padrões e Tendências:** Descobrir comportamentos, sazonalidade ou ciclos nos dados.
*   **Guiar a Pré-processamento:** Fornecer insights para a engenharia de features, tratamento de valores ausentes e limpeza de dados.
