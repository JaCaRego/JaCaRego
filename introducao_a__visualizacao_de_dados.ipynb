# INTRODUÇÃO A VISUALIZÇÃO DE DADOS

O código, foca na análise exploratória inicial dos dados de clientes, focando na distribuição de salários e na relação entre salário e anos de experiência.

**Objetivo:**

*   Carregar e exibir as primeiras linhas do conjunto de dados `clientes-v3-preparado.csv`.
*   Visualizar a distribuição do 'salario' através de histogramas, tanto básicos quanto personalizados com parâmetros adicionais.
*   Explorar a relação entre 'salario' e 'anos_experiencia' utilizando gráficos de dispersão.
*   Calcular e visualizar a correlação entre 'salario' e 'anos_experiencia' usando um mapa de calor.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('clientes-v3-preparado.csv')
print(df.head().to_string())

# Histograma
plt.hist(df['salario'])
plt.show()

# Histograma - Parâmetros
plt.figure(figsize = (10, 6))
plt.hist(df['salario'], bins = 100, color = 'green', alpha = 0.8)
plt.title('Histograma - Distribuição de salários')
plt.xlabel('Salário')
plt.xticks(ticks=range(0, int(df['salario'].max())+2000, 2000))
plt.ylabel('Frequência')
plt.grid(True)
plt.show()

# Múltiplos gráficos
plt.figure(figsize = (10, 6))
plt.subplot(2,1,1) # 2 Linha, 2 colunas, 1º Gráfico
# Gráfico de Dispersão
plt.scatter(df['salario'], df['salario'])
plt.title('Dispersão - Salario e Salário')
plt.xlabel('Salário')
plt.ylabel('Salário')

plt.subplot(1,2,2) # 1 Linha, 2 Colunas, 2º Gráfico
plt.scatter(df['salario'], df['anos_experiencia'], color = '#5883a8', alpha = 0.6, s=30) # cor hexadecimal online
plt.title('Dispersão - Idade e anos experiência')
plt.xlabel('Salário')
plt.ylabel('Anos experiência')

# Mapa de Calor

corr = df[['salario', 'anos_experiencia']].corr()
plt.subplot(2, 2, 3) # 2 linhas, 2 Colunas, 3º Gráfico
sns.heatmap(corr, annot=True, cmap = 'coolwarm')
plt.title('Correlação Salário e Idade')

plt.tight_layout() # Ajustar espaçamentos
plt.show()
````

*   **`plt.show()`**: Exibe o gráfico gerado. Sem esta linha, o gráfico pode não aparecer ou aparecer como um objeto de gráfico em vez de uma imagem.

*   **`# Histograma - Parâmetros`**: Comentário indicando um histograma com parâmetros personalizados.
*   **`plt.figure(figsize = (10, 6))`**: Cria uma nova figura para o gráfico com um tamanho específico (10 polegadas de largura por 6 polegadas de altura).
*   **`plt.hist(df['salario'], bins = 100, color = 'green', alpha = 0.8)`**: Cria o histograma com as seguintes personalizações:
    *   `bins = 100`: Define o número de barras (intervalos) no histograma para 100.
    *   `color = 'green'`: Define a cor das barras como verde.
    *   `alpha = 0.8`: Define a transparência das barras (0.8 significa 80% opaco).
*   **`plt.title('Histograma - Distribuição de salários')`**: Define o título do gráfico.
*   **`plt.xlabel('Salário')`**: Define o rótulo do eixo X.
*   **`plt.xticks(ticks=range(0, int(df['salario'].max())+2000, 2000))`**: Personaliza os marcadores (ticks) no eixo X, criando-os em intervalos de 2000, começando de 0 até um pouco além do salário máximo.
*   **`plt.ylabel('Frequência')`**: Define o rótulo do eixo Y.
*   **`plt.grid(True)`**: Adiciona uma grade ao gráfico para facilitar a leitura dos valores.
*   **`plt.show()`**: Exibe o histograma personalizado.

```python
# Múltiplos gráficos
plt.figure(figsize = (10, 6))
plt.subplot(2,1,1) # 2 Linha, 2 colunas, 1º Gráfico

plt.title('Dispersão - Idade e anos experiência')
plt.xlabel('Salário')
plt.ylabel('Anos experiência')
```
*   **`# Múltiplos gráficos`**: Comentário indicando a seção de múltiplos gráficos.
*   **`plt.figure(figsize = (10, 6))`**: Cria uma nova figura com um tamanho especificado para conter os subgráficos.
*   **`plt.subplot(2,1,1)`**: Cria uma grade de subgráficos de 2 linhas por 1 coluna e seleciona a primeira posição (superior).
*   **`# Gráfico de Dispersão`**: Comentário.
*   **`plt.scatter(df['salario'], df['salario'])`**: Cria um gráfico de dispersão com 'salario' no eixo X e 'salario' no eixo Y. Este é um gráfico redundante, pois mostra uma linha reta diagonal perfeita.
*   **`plt.title('Dispersão - Salario e Salário')`**: Define o título para o primeiro subgráfico.
*   **`plt.xlabel('Salário')`**: Define o rótulo do eixo X para o primeiro subgráfico.
*   **`plt.ylabel('Salário')`**: Define o rótulo do eixo Y para o primeiro subgráfico.
*   **`plt.subplot(1,2,2)`**: Cria uma grade de subgráficos de 1 linha por 2 colunas e seleciona a segunda posição (direita). É importante notar que o layout de subplots está sendo redefinido, o que pode causar sobreposição se não for tratado cuidadosamente (como `tight_layout` posteriormente).
*   **`plt.scatter(df['salario'], df['anos_experiencia'], color = '#5883a8', alpha = 0.6, s=30)`**: Cria um gráfico de dispersão da relação entre 'salario' e 'anos_experiencia' com personalizações:
    *   `color = '#5883a8'`: Define a cor dos pontos usando um código hexadecimal.
    *   `alpha = 0.6`: Define a transparência dos pontos.
    *   `s=30`: Define o tamanho dos pontos.
*   **`plt.title('Dispersão - Idade e anos experiência')`**: Define o título para o segundo subgráfico.
*   **`plt.xlabel('Salário')`**: Define o rótulo do eixo X para o segundo subgráfico.
*   **`plt.ylabel('Anos experiência')`**: Define o rótulo do eixo Y para o segundo subgráfico.

```python
# Mapa de Calor

corr = df[['salario', 'anos_experiencia']].corr()
plt.subplot(2, 2, 3) # 2 linhas, 2 Colunas, 3º Gráfico
sns.heatmap(corr, annot=True, cmap = 'coolwarm')
plt.title('Correlação Salário e Idade')

plt.tight_layout() # Ajustar espaçamentos
plt.show()
```
*   **`# Mapa de Calor`**: Comentário indicando a seção do mapa de calor.
*   **`corr = df[['salario', 'anos_experiencia']].corr()`**: Calcula a matriz de correlação entre as colunas 'salario' e 'anos_experiencia' do DataFrame `df`. O resultado é armazenado na variável `corr`.
*   **`plt.subplot(2, 2, 3)`**: Cria uma grade de subgráficos de 2 linhas por 2 colunas e seleciona a terceira posição. Isso posiciona o mapa de calor em um local específico na figura.
*   **`sns.heatmap(corr, annot=True, cmap = 'coolwarm')`**: Cria um mapa de calor da matriz de correlação `corr` usando `seaborn`:
    *   `annot=True`: Exibe os valores de correlação em cada célula do mapa de calor.
    *   `cmap = 'coolwarm'`: Define o mapa de cores para o gradiente de cores 'coolwarm', que é bom para correlações (cores diferentes para correlação positiva e negativa).
*   **`plt.title('Correlação Salário e Idade')`**: Define o título para o mapa de calor.
*   **`plt.tight_layout()`**: Ajusta automaticamente os parâmetros do subplot para que os elementos do gráfico (como títulos, rótulos) não se sobreponham.
*   **`plt.show()`**: Exibe todos os gráficos gerados na figura atual (os subplots).
